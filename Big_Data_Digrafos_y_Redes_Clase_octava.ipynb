```python
import networkx as nx
import matplotlib.pyplot as plt

# Crear el digrafo
D = nx.DiGraph()

# Definir los nodos (vértices)
vertices = ['a', 'b', 'c', 'd', 'e']

# Añadir los nodos al grafo
D.add_nodes_from(vertices)

# Definir los arcos (edges) según el ejemplo
edges = [('a', 'b'), ('a', 'c'), ('a', 'e'),
         ('b', 'c'), ('b', 'd'),
         ('c', 'b'), ('c', 'e'),
         ('e', 'd')]

# Añadir los arcos al grafo
D.add_edges_from(edges)

# Función para identificar arcos simétricos
def encontrar_arcos_simetricos(grafo):
    arcos_simetricos = []
    for u, v in grafo.edges():
        if grafo.has_edge(v, u):
            arcos_simetricos.append((u, v))  # Añadir si ambos arcos existen
    return arcos_simetricos

# Dibujar el grafo
pos = nx.spring_layout(D)  # Posiciones de los nodos

# Dibujar nodos y arcos
nx.draw(D, pos, with_labels=True, node_color='lightblue', node_size=2000, font_size=16, font_weight='bold', arrows=True)

# Dibujar etiquetas de los arcos (dirección de las aristas)
labels = {edge: f"{edge[0]}->{edge[1]}" for edge in D.edges}
nx.draw_networkx_edge_labels(D, pos, edge_labels=labels)

# Mostrar el gráfico
plt.title("Grafo Dirigido del Ejemplo 10.1")
plt.show()

# Cálculo de grados de salida y entrada
print("Grado de salida (outdegree) y entrada (indegree) de cada vértice:")
for v in vertices:
    print(f"Vértice {v}: outdegree = {D.out_degree(v)}, indegree = {D.in_degree(v)}")

# Encontrar e imprimir los arcos simétricos
arcos_simetricos = encontrar_arcos_simetricos(D)
print("\nArcos simétricos:")
for arco in arcos_simetricos:
    print(f"{arco[0]} <--> {arco[1]}")
```

```python
from IPython.display import display, HTML

# Contenido HTML con el Teorema 10.1 y demostración
html_content = """
<!DOCTYPE html>
<html lang="es">
<head>
    <meta charset="UTF-8">
    <meta name="viewport" content="width=device-width, initial-scale=1.0">
    <title>Teorema 10.1: Grafos Dirigidos</title>
    <style>
        body {
            font-family: 'Arial', sans-serif;
            line-height: 1.8;
            margin: 20px;
            padding: 20px;
            background-color: #f9f9f9;
        }
        h1, h2, h3 {
            color: #2c3e50;
        }
        p {
            font-size: 1.2em;
            margin-bottom: 15px;
        }
        ul {
            list-style-type: square;
            margin-left: 30px;
            margin-bottom: 20px;
        }
        li {
            margin-bottom: 10px;
            font-size: 1.1em;
        }
        .math {
            background: #f4f4f4;
            padding: 10px;
            border-radius: 5px;
            font-family: 'Cambria Math', serif;
            font-size: 1.1em;
        }
    </style>
</head>
<body>
    <h1>Teorema 10.1: Grafos Dirigidos</h1>

    <p>El <strong>Teorema 10.1</strong> que aparece en la imagen es un resultado fundamental en grafos dirigidos (o digrafos). Este teorema relaciona el número total de arcos en un digrafo con los grados de entrada (<strong>id(v)</strong>) y grados de salida (<strong>od(v)</strong>) de los vértices.</p>

    <h2>Teorema 10.1</h2>
    <p>Si <strong>D</strong> es un digrafo con:</p>
    <ul>
        <li>Conjunto de vértices <strong>V(D) = {v<sub>1</sub>, v<sub>2</sub>, …, v<sub>n</sub>}</strong> (es decir, el digrafo tiene <strong>n</strong> vértices),</li>
        <li>Número total de arcos <strong>q</strong> (es decir, las conexiones dirigidas entre vértices),</li>
    </ul>
    <p>Entonces se cumple que:</p>
    <div class="math">
        <p>∑<sub>i=1</sub><sup>n</sup> id(v<sub>i</sub>) = q</p>
        <p>y</p>
        <p>∑<sub>i=1</sub><sup>n</sup> od(v<sub>i</sub>) = q</p>
    </div>

    <h3>Interpretación del Teorema</h3>
    <p><strong>Grado de entrada total (∑ id(v<sub>i</sub>)):</strong> La suma de los grados de entrada de todos los vértices del grafo es igual al número total de arcos <strong>q</strong>.</p>
    <p>Es decir, cada arco que entra en un vértice <strong>v<sub>i</sub></strong> contribuye a su grado de entrada <strong>id(v<sub>i</sub>)</strong>, y la suma de los grados de entrada de todos los vértices debe ser igual a <strong>q</strong>, el número total de arcos en el digrafo.</p>

    <p><strong>Grado de salida total (∑ od(v<sub>i</sub>)):</strong> La suma de los grados de salida de todos los vértices también es igual al número total de arcos <strong>q</strong>.</p>
    <p>Cada arco que sale de un vértice <strong>v<sub>i</sub></strong> contribuye a su grado de salida <strong>od(v<sub>i</sub>)</strong>, y la suma de los grados de salida de todos los vértices debe ser igual a <strong>q</strong>, el número total de arcos en el digrafo.</p>

    <h3>Demostración (Proof)</h3>
    <p>Cada arco en un digrafo contribuye exactamente a dos vértices:</p>
    <ul>
        <li>Al grado de entrada de un vértice: el vértice hacia donde apunta el arco.</li>
        <li>Al grado de salida de otro vértice: el vértice desde donde sale el arco.</li>
    </ul>
    <p>Por lo tanto, cada arco incrementa en 1 tanto el grado de salida de un vértice como el grado de entrada de otro vértice. Como consecuencia:</p>
    <p class="math">La suma total de los grados de entrada y salida de todos los vértices es igual al número total de arcos q.</p>

    <h3>Ejemplo</h3>
    <p>Supongamos que tenemos un digrafo con 4 vértices <strong>V(D) = {a, b, c, d}</strong> y 5 arcos:</p>
    <div class="math">
        <p>E(D) = {(a, b), (a, c), (b, d), (c, b), (d, a)}</p>
    </div>

    <h3>Grados de entrada id(v)</h3>
    <ul>
        <li><strong>id(a) = 1</strong> (el arco (d, a))</li>
        <li><strong>id(b) = 2</strong> (los arcos (a, b) y (c, b))</li>
        <li><strong>id(c) = 1</strong> (el arco (a, c))</li>
        <li><strong>id(d) = 1</strong> (el arco (b, d))</li>
    </ul>
    <p>Suma de los grados de entrada: <strong>1 + 2 + 1 + 1 = 5</strong>.</p>

    <h3>Grados de salida od(v)</h3>
    <ul>
        <li><strong>od(a) = 2</strong> (los arcos (a, b) y (a, c))</li>
        <li><strong>od(b) = 1</strong> (el arco (b, d))</li>
        <li><strong>od(c) = 1</strong> (el arco (c, b))</li>
        <li><strong>od(d) = 1</strong> (el arco (d, a))</li>
    </ul>
    <p>Suma de los grados de salida: <strong>2 + 1 + 1 + 1 = 5</strong>.</p>

    <p>Tanto la suma de los grados de entrada como la de los grados de salida es <strong>5</strong>, que es igual al número total de arcos en el digrafo (q = 5).</p>

</body>
</html>
"""

# Mostrar el contenido HTML en Colab
display(HTML(html_content))
```

```python
import networkx as nx
import matplotlib.pyplot as plt

# Función para crear un digrafo y verificar el Teorema 10.1
def verificar_teorema_10_1(arcos):
    # Crear el digrafo
    D = nx.DiGraph()

    # Añadir arcos al grafo
    D.add_edges_from(arcos)

    # Calcular grados de entrada y salida
    grados_entrada = [D.in_degree(v) for v in D.nodes()]
    grados_salida = [D.out_degree(v) for v in D.nodes()]

    # Calcular el número total de arcos
    q = D.number_of_edges()

    # Verificar el teorema
    suma_grados_entrada = sum(grados_entrada)
    suma_grados_salida = sum(grados_salida)

    print("Suma de los grados de entrada:", suma_grados_entrada)
    print("Suma de los grados de salida:", suma_grados_salida)
    print("Número total de arcos (q):", q)

    # Verificar si se cumple el teorema
    if suma_grados_entrada == q and suma_grados_salida == q:
        print("Teorema 10.1 verificado: La suma de grados de entrada y salida es igual al número de arcos.")
    else:
        print("El teorema 10.1 no se cumple.")

    # Dibujar el grafo
    pos = nx.spring_layout(D)
    nx.draw(D, pos, with_labels=True, node_size=2000, node_color='lightblue', font_size=16, arrows=True)
    plt.show()

# Ejemplo 1: Grafo dirigido pequeño
arcos_ejemplo_1 = [('a', 'b'), ('b', 'c'), ('c', 'a')]
print("Ejemplo 1:")
verificar_teorema_10_1(arcos_ejemplo_1)

# Ejemplo 2: Grafo dirigido más grande
arcos_ejemplo_2 = [('a', 'b'), ('a', 'c'), ('b', 'd'), ('c', 'd')]
print("\nEjemplo 2:")
verificar_teorema_10_1(arcos_ejemplo_2)
```

```python
import networkx as nx
import matplotlib.pyplot as plt

def es_fuertemente_conectado(grafo):
    """ Verifica si el grafo dirigido es fuertemente conectado """
    return nx.is_strongly_connected(grafo)

def tiene_puentes(grafo):
    """ Verifica si el grafo no dirigido tiene puentes """
    return len(list(nx.bridges(grafo))) > 0

def orientar_grafo(grafo):
    """ Orienta un grafo no dirigido sin puentes para hacerlo fuertemente conectado """
    if tiene_puentes(grafo):
        print("El grafo tiene puentes, no puede ser orientado para ser fuertemente conectado.")
        return None

    # Convertir el grafo no dirigido en un grafo dirigido usando DFS
    dirigido = nx.DiGraph()

    # Realizamos DFS y orientamos las aristas
    dfs_edges = list(nx.dfs_edges(grafo))
    dirigido.add_edges_from(dfs_edges)  # Orientamos hacia los nodos alcanzados en el DFS

    # Ahora orientamos las aristas restantes
    for u, v in grafo.edges():
        if not dirigido.has_edge(u, v) and not dirigido.has_edge(v, u):
            # Orientamos las aristas que no están en el DFS, hacia el nodo con menor número
            if u < v:
                dirigido.add_edge(v, u)
            else:
                dirigido.add_edge(u, v)

    return dirigido

def dibujar_grafo(grafo, titulo):
    """ Dibuja el grafo con etiquetas y direcciones de las aristas """
    pos = nx.spring_layout(grafo)
    plt.figure(figsize=(8, 6))
    nx.draw(grafo, pos, with_labels=True, node_color='lightblue', node_size=2000, font_size=16, font_weight='bold', arrows=True)
    edge_labels = {(u, v): f"{u}->{v}" for u, v in grafo.edges()}
    nx.draw_networkx_edge_labels(grafo, pos, edge_labels=edge_labels)
    plt.title(titulo)
    plt.show()

# Ejemplo 1: Grafo sin puentes
grafo1 = nx.Graph()
grafo1.add_edges_from([(1, 2), (2, 3), (3, 4), (4, 1), (2, 4)])

# Verificar si tiene puentes
print("Ejemplo 1: Grafo sin puentes")
if tiene_puentes(grafo1):
    print("El grafo tiene puentes.")
else:
    print("El grafo no tiene puentes.")
    # Orientar el grafo
    grafo_orientado1 = orientar_grafo(grafo1)
    if grafo_orientado1:
        dibujar_grafo(grafo_orientado1, "Grafo orientado (fuertemente conectado)")

# Ejemplo 2: Grafo con un puente
grafo2 = nx.Graph()
grafo2.add_edges_from([(1, 2), (2, 3), (3, 4), (4, 5)])  # El borde (4, 5) es un puente

print("\nEjemplo 2: Grafo con puente")
if tiene_puentes(grafo2):
    print("El grafo tiene puentes.")
else:
    print("El grafo no tiene puentes.")
    # Orientar el grafo
    grafo_orientado2 = orientar_grafo(grafo2)
    if grafo_orientado2:
        dibujar_grafo(grafo_orientado2, "Grafo orientado (fuertemente conectado)")
```

```python
import networkx as nx
import matplotlib.pyplot as plt

def crear_dag():
    """ Crea un DAG (grafo dirigido acíclico) que representa dependencias de tareas """
    dag = nx.DiGraph()

    # Añadimos las tareas como nodos
    dag.add_nodes_from(['Tarea 1', 'Tarea 2', 'Tarea 3', 'Tarea 4', 'Tarea 5', 'Tarea 6'])

    # Añadimos las dependencias entre las tareas como aristas
    # Ejemplo: 'Tarea 1' debe completarse antes que 'Tarea 2', por eso hay una arista 'Tarea 1' -> 'Tarea 2'
    dag.add_edges_from([
        ('Tarea 1', 'Tarea 2'),
        ('Tarea 1', 'Tarea 3'),
        ('Tarea 2', 'Tarea 4'),
        ('Tarea 3', 'Tarea 4'),
        ('Tarea 4', 'Tarea 5'),
        ('Tarea 4', 'Tarea 6')
    ])

    return dag

def dibujar_dag(dag, titulo):
    """ Dibuja el DAG con las tareas y sus dependencias """
    plt.figure(figsize=(10, 7))

    # Distribución con más espacio entre nodos
    pos = nx.spring_layout(dag, k=1.5, iterations=50)  # k controla la repulsión, mayor k separa más los nodos
    nx.draw(dag, pos, with_labels=True, node_size=3000, node_color='lightblue', font_size=14, font_weight='bold', arrows=True, arrowstyle='->', arrowsize=20)

    # Mostrar etiquetas de las aristas (dependencias)
    edge_labels = {(u, v): f"{u}->{v}" for u, v in dag.edges()}
    nx.draw_networkx_edge_labels(dag, pos, edge_labels=edge_labels, font_size=10)

    plt.title(titulo)
    plt.show()

def ordenacion_topologica(dag):
    """ Realiza una ordenación topológica de las tareas en el DAG """
    try:
        orden = list(nx.topological_sort(dag))
        print("Ordenación topológica de las tareas:")
        for i, tarea in enumerate(orden, 1):
            print(f"{i}. {tarea}")
        return orden
    except nx.NetworkXUnfeasible:
        print("El grafo no es un DAG (hay ciclos presentes).")

def verificar_acyclicidad(dag):
    """ Verifica si el grafo es acíclico (DAG) """
    if nx.is_directed_acyclic_graph(dag):
        print("El grafo es un DAG (acíclico).")
    else:
        print("El grafo no es un DAG (hay ciclos).")

# Creación del DAG
dag = crear_dag()

# Dibujar el DAG con una mejor separación
dibujar_dag(dag, "Dependencias de Tareas en un DAG")

# Verificar si es acíclico
verificar_acyclicidad(dag)

# Realizar la ordenación topológica
orden_topologico = ordenacion_topologica(dag)
```
