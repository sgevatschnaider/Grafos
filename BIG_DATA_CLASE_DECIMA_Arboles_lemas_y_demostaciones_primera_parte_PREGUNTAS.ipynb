```python
from IPython.display import display, HTML

# Contenido HTML con el enunciado del Lema 3.1.1
html_content = """
<!DOCTYPE html>
<html lang="es">
<head>
    <meta charset="UTF-8">
    <meta name="viewport" content="width=device-width, initial-scale=1.0">
    <title>Enunciado del Lema 3.1.1</title>
    <link href="https://fonts.googleapis.com/css2?family=Roboto:wght@400;700&display=swap" rel="stylesheet">
    <style>
        body {
            font-family: 'Roboto', Arial, sans-serif;
            line-height: 1.8;
            margin: 20px;
            padding: 20px;
            background-color: #f9f9f9;
            color: #333;
        }
        h1 {
            font-size: 2.5em;
            text-align: center;
            color: #2c3e50;
            margin-bottom: 20px;
        }
        h2 {
            color: #2980b9;
            margin-top: 30px;
        }
        p, li {
            font-size: 1.2em;
            margin-bottom: 10px;
        }
        ul {
            list-style-type: square;
            margin-left: 20px;
        }
        code {
            background-color: #f0f0f0;
            padding: 2px 5px;
            border-radius: 3px;
        }
        .example {
            background-color: #e8f8f5;
            border-left: 5px solid #1abc9c;
            padding: 15px;
            margin-top: 20px;
        }
        .highlight {
            background-color: #ffeb3b;
            padding: 2px 5px;
            border-radius: 3px;
            font-weight: bold;
        }
    </style>
</head>
<body>

    <h1>Enunciado del Lema 3.1.1</h1>

    <p><strong>"Una arista <code>e ∈ E(G)</code> es un puente de <code>G</code> si y solo si <code>e</code> no está contenida en ningún ciclo en <code>G</code>."</strong></p>

    <h2>Conceptos Clave</h2>

    <ul>
        <li><strong>Grafo <code>G</code>:</strong> <code>G</code> es un grafo que contiene un conjunto de vértices <code>V(G)</code> y un conjunto de aristas <code>E(G)</code>.</li>
        <li><strong>Puente:</strong> Una arista <code>e</code> en un grafo <code>G</code> se llama puente si, al eliminarla, el grafo <code>G</code> se desconecta, o si la componente conectada que contenía <code>e</code> se divide en dos componentes separadas. Es decir, el número de componentes conectadas en <code>G</code> aumenta al eliminar <code>e</code>.</li>
        <li><strong>Ciclo:</strong> Un ciclo en un grafo es una secuencia cerrada de aristas y vértices en la cual cada vértice se toca solo una vez (salvo el inicio y el fin, que coinciden). Un ciclo no repite aristas y pasa solo una vez por cada vértice intermedio.</li>
    </ul>

    <h2>Explicación del Lema</h2>

    <p>El lema afirma que una arista <code>e</code> en <code>G</code> es un puente si y solo si no forma parte de ningún ciclo en <code>G</code>. Esto implica dos cosas:</p>

    <ul>
        <li><strong>(Si)</strong> <code>e</code> es un puente, entonces no existe un ciclo que la contenga.</li>
        <li><strong>(Solo si)</strong> <code>e</code> no forma parte de ningún ciclo, entonces es un puente.</li>
    </ul>

    <h3>(Si) <code>e</code> es un puente ⇒ <code>e</code> no está en ningún ciclo</h3>
    <p>Supongamos que <code>e</code> es un puente en <code>G</code>. Si <code>e</code> fuera parte de un ciclo, podríamos recorrer el ciclo sin importar si eliminamos <code>e</code> o no. Esto significaría que <code>G</code> sigue conectado incluso sin <code>e</code>, lo cual contradice la definición de puente. Por lo tanto, si <code>e</code> es un puente, no puede estar en ningún ciclo.</p>

    <h3>(Solo si) <code>e</code> no está en ningún ciclo ⇒ <code>e</code> es un puente</h3>
    <p>Si <code>e</code> no está contenida en ningún ciclo en <code>G</code>, no hay otra forma de conectar los dos extremos de <code>e</code> mediante otras aristas o rutas en <code>G</code>. Al remover <code>e</code>, el grafo se desconecta, porque los extremos de <code>e</code> ya no tienen ninguna ruta alternativa para conectarse. Así, <code>e</code> actúa como un puente en <code>G</code>.</p>

    <h2>Conclusión</h2>
    <p>El lema establece una equivalencia: una arista es un puente si y solo si no pertenece a ningún ciclo. Esto nos proporciona una herramienta útil para identificar puentes en un grafo: basta con verificar si una arista es parte de algún ciclo para determinar si es un puente.</p>

    <div class="example">
        <h2>Ejemplo</h2>
        <p>Considera un grafo simple <code>G</code> con los vértices <code>V = {A, B, C, D}</code> y las aristas <code>E = {AB, BC, CD, DA, AC}</code>. Este grafo contiene un ciclo, por ejemplo, <code>A → B → C → A</code>.</p>
        <p>Ahora:</p>
        <ul>
            <li>Si añadimos una arista <code>BE</code> conectando <code>B</code> y un vértice adicional <code>E</code>, y <code>BE</code> no pertenece a ningún ciclo, entonces <code>BE</code> es un puente.</li>
            <li>Cualquier arista que forme parte de algún ciclo, como <code>AB</code>, no será un puente, ya que existen otras rutas que conectan los vértices si removemos <code>AB</code>.</li>
        </ul>
    </div>

</body>
</html>
"""

# Mostrar el contenido HTML en Colab
display(HTML(html_content))
```

```python
import networkx as nx
import matplotlib.pyplot as plt

# Crear un árbol con 6 vértices (conexo y acíclico)
G = nx.Graph()
edges = [(0, 1), (1, 2), (1, 3), (3, 4), (3, 5)]
G.add_edges_from(edges)

# Posiciones de los nodos para un layout limpio
pos = nx.spring_layout(G, seed=42)

# Función para graficar el árbol
def draw_graph_with_edges(G, pos, title):
    plt.figure(figsize=(8, 6))
    nx.draw(G, pos, with_labels=True, node_color="lightblue", edge_color="gray", node_size=700, font_size=12)
    plt.title(title)
    plt.show()

# 1. Graficar el árbol original
draw_graph_with_edges(G, pos, "Grafo Conexo y Acíclico (Un Árbol)")

# 2. Verificar que todas las aristas son puentes
def check_if_bridge(G, edge):
    # Eliminar la arista
    G_copy = G.copy()
    G_copy.remove_edge(*edge)
    # Verificar si el grafo se desconecta
    return not nx.is_connected(G_copy)

# Mostrar cuáles aristas son puentes
print("Aristas en el grafo original:")
for edge in G.edges():
    is_bridge = check_if_bridge(G, edge)
    print(f"Arista {edge} es un puente: {is_bridge}")

# 3. Verificar que tiene exactamente |V(G)| - 1 aristas
num_vertices = G.number_of_nodes()
num_edges = G.number_of_edges()

print(f"\nNúmero de vértices: {num_vertices}")
print(f"Número de aristas: {num_edges}")
print(f"¿Tiene exactamente |V(G)| - 1 aristas? {'Sí' if num_edges == num_vertices - 1 else 'No'}")

# 4. Eliminar una arista y mostrar cómo se desconecta el grafo
G_removed = G.copy()
G_removed.remove_edge(1, 3)

# Graficar el grafo después de eliminar una arista (puente)
draw_graph_with_edges(G_removed, pos, "Grafo después de eliminar una arista (puente)")
```

```python
from IPython.display import display, HTML

# Contenido HTML sobre la construcción de un árbol generador
html_content = """
<!DOCTYPE html>
<html lang="es">
<head>
    <meta charset="UTF-8">
    <meta name="viewport" content="width=device-width, initial-scale=1.0">
    <title>Construcción de un Árbol Generador</title>
    <link href="https://fonts.googleapis.com/css2?family=Roboto:wght@400;700&display=swap" rel="stylesheet">
    <style>
        body {
            font-family: 'Roboto', Arial, sans-serif;
            line-height: 1.8;
            margin: 20px;
            padding: 20px;
            background-color: #f9f9f9;
            color: #333;
        }
        h1 {
            font-size: 2.5em;
            text-align: center;
            color: #2c3e50;
            margin-bottom: 20px;
        }
        h2 {
            color: #2980b9;
            margin-top: 30px;
        }
        p, li {
            font-size: 1.2em;
            margin-bottom: 10px;
        }
        ul {
            list-style-type: square;
            margin-left: 20px;
        }
        code {
            background-color: #f0f0f0;
            padding: 2px 5px;
            border-radius: 3px;
        }
        .highlight {
            background-color: #ffeb3b;
            padding: 2px 5px;
            border-radius: 3px;
            font-weight: bold;
        }
        .example {
            background-color: #e8f8f5;
            border-left: 5px solid #1abc9c;
            padding: 15px;
            margin-top: 20px;
        }
    </style>
</head>
<body>

    <h1>Construcción de un Árbol Generador</h1>

    <h2>Objetivo de la Última Parte</h2>
    <p>Esta parte de la demostración trata de construir un <strong>árbol generador</strong> de un grafo conexo <code>G</code>, si <code>G</code> tiene ciclos. Un <strong>árbol generador</strong> es un subgrafo de <code>G</code> que incluye todos los vértices de <code>G</code>, pero sin ciclos, y sigue siendo conexo.</p>

    <h2>Proceso de Construcción del Árbol Generador</h2>
    <ul>
        <li><strong>Partimos de un grafo <code>G</code> conexo que contiene ciclos:</strong>
        <p><code>G</code> es un grafo donde todos sus vértices están conectados, pero puede haber ciclos.</p></li>

        <li><strong>Identificación y eliminación de ciclos:</strong>
        <p>Mientras exista un ciclo en <code>G</code>, elegimos una arista cualquiera de ese ciclo y la <strong>eliminamos</strong>.
        Dado que el ciclo tiene múltiples caminos que conectan sus vértices, eliminar una arista del ciclo <strong>no desconecta el grafo</strong>; simplemente elimina un camino redundante.</p></li>

        <li><strong>Aplicación del Lema 3.1.1:</strong>
        <p>El <strong>Lema 3.1.1</strong> establece que una arista es un <strong>puente</strong> si y solo si no pertenece a ningún ciclo. Al eliminar aristas de los ciclos, no eliminamos puentes, por lo tanto, <code>G</code> permanece <strong>conexo</strong>.</p></li>

        <li><strong>Repetición hasta eliminar todos los ciclos:</strong>
        <p>Eliminamos aristas de los ciclos hasta que ya no haya ciclos en el grafo. El subgrafo resultante es:</p>
        <ul>
            <li><strong>Conexo:</strong> Nunca eliminamos puentes, solo aristas de los ciclos.</li>
            <li><strong>Acíclico:</strong> Ya no quedan aristas que formen ciclos.</li>
        </ul>
        </li>

        <li><strong>Subgrafo resultante: Árbol generador</strong>
        <p>El subgrafo conexo y acíclico obtenido al final incluye todos los vértices de <code>G</code>, lo que lo convierte en un <strong>árbol generador</strong> o <strong>spanning tree</strong> de <code>G</code>. Este árbol generador utiliza exactamente <code>|V(G)| - 1</code> aristas.</p></li>
    </ul>

    <h2>Conclusión</h2>
    <p>La última parte de la demostración muestra que cualquier grafo conexo <code>G</code> tiene un subgrafo conexo y acíclico que abarca todos sus vértices: un <strong>árbol generador</strong>. Para obtenerlo:</p>
    <ul>
        <li>Identificamos y eliminamos aristas en ciclos, uno a uno, sin desconectar el grafo, hasta que queda un subgrafo acíclico.</li>
    </ul>
    <p>Este árbol generador es un concepto fundamental en teoría de grafos, ya que proporciona una estructura acíclica mínima que conserva la conectividad del grafo original <code>G</code>.</p>

</body>
</html>
"""

# Mostrar el contenido HTML en Google Colab
display(HTML(html_content))
```

```python
import networkx as nx
import matplotlib.pyplot as plt

# Crear un grafo con ciclos
G = nx.Graph()
edges_with_cycles = [(0, 1), (1, 2), (2, 3), (3, 0), (1, 3), (3, 4), (4, 5), (5, 6), (6, 4)]  # Grafo con ciclos
G.add_edges_from(edges_with_cycles)

# Posiciones de los nodos para la visualización
pos = nx.spring_layout(G, seed=42)

# Función para graficar el grafo
def draw_graph(G, pos, title, highlight_edges=[]):
    plt.figure(figsize=(8, 6))
    nx.draw(G, pos, with_labels=True, node_color="lightblue", edge_color="gray", node_size=700, font_size=12)

    # Resaltar aristas si están en el árbol generador
    if highlight_edges:
        nx.draw_networkx_edges(G, pos, edgelist=highlight_edges, edge_color='red', width=2.5)

    plt.title(title)
    plt.show()

# 1. Graficar el grafo original con ciclos
draw_graph(G, pos, "Grafo Con Ciclos")

# 2. Eliminar ciclos para obtener el árbol generador
tree_edges = list(nx.minimum_spanning_edges(G, algorithm="kruskal", data=False))  # Algoritmo de Kruskal para eliminar ciclos
tree_G = nx.Graph()
tree_G.add_edges_from(tree_edges)

# 3. Graficar el árbol generador (spanning tree)
draw_graph(G, pos, "Árbol Generador (sin ciclos)", highlight_edges=tree_edges)

# Mostrar el proceso de eliminación de ciclos
print("Aristas originales del grafo con ciclos:", edges_with_cycles)
print("Aristas seleccionadas para el árbol generador:", tree_edges)
```

 **Preguntas**

### 1. **¿Qué es un puente en un grafo?**
   **Respuesta**: Un puente es una arista en un grafo que, si se elimina, desconecta el grafo. Esto significa que el número de componentes conectadas del grafo aumenta al eliminar esa arista. En otras palabras, un puente es crucial para mantener la conectividad de un grafo. El **Lema 3.1.1** establece que una arista es un puente si y solo si no está contenida en ningún ciclo del grafo.

---

### 2. **¿Qué establece el Lema 3.1.1?**
   **Respuesta**: El Lema 3.1.1 establece que una arista \( e \in E(G) \) es un puente en un grafo \( G \) si y solo si \( e \) no está contenida en ningún ciclo de \( G \). Es decir, si una arista pertenece a un ciclo, no puede ser un puente, porque al eliminarla, el grafo sigue siendo conexo. Por otro lado, si una arista no pertenece a ningún ciclo, entonces es un puente.

---

### 3. **¿Qué condiciones son equivalentes a que un grafo \( G \) sea un árbol según la Proposición 3.1.2?**
   **Respuesta**: La Proposición 3.1.2 establece tres condiciones equivalentes para que un grafo \( G \) sea un árbol:
   1. \( G \) es conexo y acíclico.
   2. \( G \) es conexo y todas sus aristas son puentes.
   3. \( G \) es conexo y tiene exactamente \( |V(G)| - 1 \) aristas.
   Estas condiciones son equivalentes entre sí, lo que significa que si una de ellas es cierta, entonces las otras también lo son.

---

### 4. **¿Cómo se relaciona la aciclicidad de un grafo con sus aristas puente?**
   **Respuesta**: Según el **Lema 3.1.1**, en un grafo acíclico, todas las aristas son puentes. Esto se debe a que en un grafo acíclico, no existen ciclos que conecten los extremos de una arista mediante rutas alternativas. Por lo tanto, eliminar cualquier arista en un grafo acíclico desconecta el grafo, convirtiendo a cada arista en un puente.

---

### 5. **¿Cómo se puede demostrar que un árbol tiene \( |V(G)| - 1 \) aristas?**
   **Respuesta**: Esta afirmación se demuestra mediante **inducción fuerte**:
   - **Caso base**: Si el grafo tiene un solo vértice, no tiene aristas, y la fórmula \( |V(G)| - 1 = 0 \) es válida.
   - **Paso inductivo**: Si asumimos que un árbol con \( n \) vértices tiene \( n - 1 \) aristas, y luego agregamos un vértice adicional conectado por una arista, obtenemos un árbol con \( n + 1 \) vértices y \( n \) aristas, lo que cumple con la fórmula \( |V(G)| - 1 \).

---

### 6. **¿Qué se entiende por un árbol generador (spanning tree)?**
   **Respuesta**: Un **árbol generador** es un subgrafo de un grafo \( G \) que:
   - Contiene todos los vértices del grafo \( G \).
   - Es conexo y acíclico.
   - Tiene \( |V(G)| - 1 \) aristas.
   Un árbol generador es esencialmente una forma de conectar todos los vértices de \( G \) sin ciclos y usando el mínimo número de aristas.

---

### 7. **¿Cómo se puede construir un árbol generador de un grafo conexo?**
   **Respuesta**: Para construir un árbol generador a partir de un grafo conexo, se puede seguir el siguiente proceso:
   1. Buscar un ciclo en el grafo.
   2. Eliminar una arista del ciclo sin desconectar el grafo (según el **Lema 3.1.1**).
   3. Repetir el proceso hasta que no queden ciclos en el grafo.
   Al final de este procedimiento, se obtiene un subgrafo acíclico y conexo, es decir, un árbol generador.

---

### 8. **¿Por qué un grafo conexo con \( |V(G)| - 1 \) aristas es acíclico?**
   **Respuesta**: Si un grafo conexo tiene \( |V(G)| - 1 \) aristas, no puede contener ciclos. Esto se debe a que un ciclo introduce redundancia, lo que implica que el grafo podría tener más aristas de las necesarias para conectar todos sus vértices. Si existiera un ciclo, podríamos eliminar una arista del ciclo sin desconectar el grafo, lo cual contradice la condición de que el grafo tiene exactamente \( |V(G)| - 1 \) aristas.

---

### 9. **¿Cómo se demuestra que eliminar una arista de un ciclo no desconecta el grafo?**
   **Respuesta**: En un ciclo, cada vértice está conectado a través de múltiples caminos. Si eliminamos una arista de un ciclo, los vértices de ese ciclo siguen conectados a través de las otras aristas que componen el ciclo. Por lo tanto, la eliminación de una sola arista de un ciclo no desconecta el grafo, ya que todavía existe al menos un camino entre todos los vértices involucrados en el ciclo.

---

### 10. **¿Qué dice la Proposición 3.1.3 sobre los grafos conexos?**
   **Respuesta**: La **Proposición 3.1.3** establece que **cualquier grafo conexo contiene un árbol generador**. Esto significa que, dado un grafo conexo, siempre es posible encontrar un subgrafo que incluya todos los vértices del grafo original y que sea acíclico. Un árbol generador conserva la conectividad del grafo y tiene exactamente \( |V(G)| - 1 \) aristas.
