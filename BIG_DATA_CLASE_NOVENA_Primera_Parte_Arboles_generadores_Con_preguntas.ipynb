```python
from IPython.core.display import display, HTML

# Código HTML mejorado con más espacio y mejor legibilidad
html_code = """
<h2 style="color: #2E8B57; line-height: 1.5;">Puentes, árboles y algoritmos</h2>

<p style="line-height: 1.8; margin-bottom: 20px;">En el contexto de <strong style="color: #4169E1;">teoría de grafos</strong>, vamos a explorar dos conceptos importantes: <strong style="color: #4169E1;">árboles</strong> y <strong style="color: #4169E1;">puentes</strong>.</p>

<h3 style="color: #8B4513; margin-top: 30px; margin-bottom: 20px;">1. Árboles y Bosques</h3>

<ul style="margin-left: 20px; line-height: 1.6; margin-bottom: 20px;">
  <li><strong>Árbol</strong>: Un grafo que es <strong style="color: #4169E1;">conexo</strong> y <strong style="color: #4169E1;">acíclico</strong>.
    <ul style="margin-left: 20px;">
      <li><strong>Conexo</strong>: Cualquier par de vértices está conectado por un camino.</li>
      <li><strong>Acíclico</strong>: No contiene ciclos (no puedes empezar en un vértice, moverte por el grafo y volver al vértice original sin repetir una arista).</li>
    </ul>
  </li>
</ul>

<p style="line-height: 1.8; margin-bottom: 20px;">Los <strong>árboles</strong> son estructuras útiles en diversas disciplinas, desde la informática (como en árboles binarios) hasta las ciencias de datos.</p>

<p style="line-height: 1.8; margin-bottom: 20px;"><strong>Bosque</strong>: Un grafo <strong style="color: #4169E1;">acíclico</strong> que puede estar desconectado. Cada componente de un bosque es un árbol.</p>

<hr style="margin-top: 40px; margin-bottom: 40px;">

<h3 style="color: #8B4513; margin-top: 30px; margin-bottom: 20px;">2. Puentes en un Grafo</h3>

<p style="line-height: 1.8; margin-bottom: 20px;">Un <strong>puente</strong> (o arista de corte) en un grafo G es una arista <em>e</em> en E(G) tal que al eliminarla del grafo, el número de componentes conexas aumenta. En otras palabras, un puente es una arista cuya eliminación <strong style="color: #FF6347;">desconecta</strong> alguna parte del grafo.</p>

<p style="line-height: 1.8; margin-bottom: 20px;"><strong>Formalmente:</strong> Si G - e (el grafo que resulta de eliminar la arista <em>e</em>) tiene más componentes que G, entonces <em>e</em> es un <strong>puente</strong>.</p>

<p style="line-height: 1.8; margin-bottom: 20px;">Por ejemplo, los bordes p<sub>1</sub>p<sub>2</sub>, p<sub>1</sub>p<sub>4</sub> y p<sub>4</sub>p<sub>3</sub> son <strong>puentes</strong> porque al eliminarlos, el grafo se desconecta. Sin embargo, otras aristas no lo son porque su eliminación no desconecta el grafo.</p>

<hr style="margin-top: 40px; margin-bottom: 40px;">

<h3 style="color: #8B4513; margin-top: 30px; margin-bottom: 20px;">3. Lema 3.1.1: Caracterización de un Puente</h3>

<p style="line-height: 1.8; margin-bottom: 20px;">El <strong>Lema 3.1.1</strong> ofrece una forma sencilla de identificar un puente:</p>

<div style="border-left: 5px solid #8B0000; padding-left: 10px; margin-left: 20px; margin-bottom: 20px;">
  <p style="line-height: 1.8; margin-bottom: 20px;"><strong>Lema:</strong> Una arista <em>e</em> en E(G) es un puente si y solo si no está contenida en ningún ciclo de G.</p>
</div>

<p style="line-height: 1.8; margin-bottom: 20px;"><strong>Demostración:</strong></p>
<ul style="line-height: 1.6; margin-bottom: 20px;">
  <li>Si <em>e</em> está en un ciclo C de G, entonces eliminar <em>e</em> no desconectaría el grafo, ya que hay otros caminos (debido al ciclo) que conectan los extremos de <em>e</em>. Así, G - e sigue siendo conexo, lo que implica que <em>e</em> no es un puente.</li>
  <li>Si <em>e</em> no está en ningún ciclo, su eliminación desconecta el grafo, por lo tanto, <em>e</em> es un puente.</li>
</ul>

<p style="line-height: 1.8; margin-bottom: 20px;">Dado que los <strong>árboles</strong> no tienen ciclos, <strong>todas las aristas de un árbol son puentes</strong>.</p>

<hr style="margin-top: 40px; margin-bottom: 40px;">

<h3 style="color: #8B4513; margin-top: 30px; margin-bottom: 20px;">4. Caracterización de los Árboles</h3>

<p style="line-height: 1.8; margin-bottom: 20px;">Los árboles se pueden caracterizar de varias maneras:</p>

<ol style="line-height: 1.6; margin-bottom: 20px;">
  <li>Un grafo conexo acíclico es un árbol.</li>
  <li>En un árbol, <strong>cada arista es un puente</strong>.</li>
  <li>Un árbol de <em>n</em> vértices siempre tiene <em>n - 1</em> aristas. Esto es porque los árboles son conexos pero no tienen ciclos, por lo que no pueden tener más aristas que el mínimo necesario para conectarlos sin redundancia.</li>
</ol>

<hr style="margin-top: 40px; margin-bottom: 40px;">

<h2 style="color: #2E8B57; margin-top: 30px; margin-bottom: 20px;">Aplicaciones y Algoritmos Relacionados</h2>

<ol style="line-height: 1.6; margin-bottom: 20px;">
  <li><strong>Detección de puentes:</strong> Los puentes son importantes en redes, ya que su eliminación puede desconectar partes de la red. Algoritmos como el de <strong style="color: #FF6347;">Tarjan</strong> permiten identificar todos los puentes de un grafo en tiempo lineal O(V + E), donde V es el número de vértices y E el número de aristas.</li>
  <li><strong>Árboles de expansión mínima:</strong> Estos árboles conectan todos los vértices de un grafo con el menor costo posible en términos de aristas. Los algoritmos de <strong>Kruskal</strong> y <strong>Prim</strong> son ejemplos famosos para encontrar árboles de expansión mínima.</li>
</ol>


"""

# Mostrar el HTML en Colab
display(HTML(html_code))
```

```python
from IPython.core.display import display, HTML

# Contenido HTML para la teoría de grafos y árboles generadores
html_content = """
<!DOCTYPE html>
<html lang="es">
<head>
    <meta charset="UTF-8">
    <meta name="viewport" content="width=device-width, initial-scale=1.0">
    <title>Árbol Generador en la Teoría de Grafos</title>
    <style>
        body {
            font-family: Arial, sans-serif;
            line-height: 1.8;
            margin: 20px;
            padding: 20px;
            background-color: #f9f9f9;
        }
        h1, h2, h3 {
            color: #2E8B57;
            margin-top: 20px;
        }
        h1 {
            font-size: 2.5em;
            text-align: center;
        }
        h2 {
            font-size: 1.8em;
        }
        p {
            font-size: 1.2em;
            margin-bottom: 15px;
        }
        ul {
            list-style-type: disc;
            margin-left: 30px;
            margin-bottom: 20px;
        }
        li {
            margin-bottom: 10px;
        }
        code {
            background-color: #f4f4f4;
            padding: 2px 4px;
            font-family: Consolas, "Courier New", monospace;
        }
        pre {
            background-color: #f4f4f4;
            padding: 10px;
            border-radius: 5px;
            font-size: 1em;
            overflow-x: auto;
        }
    </style>
</head>
<body>

    <h1>Árbol Generador en la Teoría de Grafos</h1>

    <p>En el campo matemático de la <strong>teoría de grafos</strong>, un <strong>árbol generador</strong> (<em>spanning tree</em> en inglés) de un grafo no dirigido <code>G</code> es un subgrafo que tiene la forma de un árbol (es decir, un grafo sin ciclos) y que incluye todos los vértices de <code>G</code>. Dicho de otra forma, un árbol generador cubre todos los nodos del grafo <code>G</code> sin formar ningún ciclo.</p>

    <h2>Conceptos Clave</h2>
    <ul>
        <li><strong>Subgrafo:</strong> Un subgrafo es un grafo que se forma tomando algunos de los vértices y aristas del grafo original. En este caso, el subgrafo es un árbol que debe conectar todos los vértices del grafo original.</li>
        <li><strong>Árbol:</strong> Un árbol es un grafo que no tiene ciclos y está conectado (es decir, hay un camino entre cualquier par de vértices).</li>
        <li><strong>Árbol generador:</strong> Es un subgrafo de <code>G</code> que incluye todos los vértices, no tiene ciclos y está conectado.</li>
    </ul>

    <h2>Ejemplos</h2>
    <p>Si tienes un grafo con 5 vértices y varias aristas, el árbol generador es una de las posibles maneras de conectar todos los vértices sin crear ningún ciclo. Si eliminas algunas aristas del grafo original de forma que aún mantienes todos los vértices conectados, pero no tienes ningún ciclo, obtendrás un árbol generador.</p>

    <h2>Propiedades de los Árboles Generadores</h2>
    <ul>
        <li><strong>Grafos conectados:</strong> Para que un grafo tenga un árbol generador, debe ser conectado. Esto significa que debe haber un camino entre cualquier par de vértices. Si el grafo no es conectado, pueden existir bosques generadores (<em>spanning forests</em>), donde cada componente conectado tiene su propio árbol generador.</li>
        <li><strong>Unicidad en los árboles:</strong> Si el grafo original <code>G</code> ya es un árbol, entonces su único árbol generador es el mismo grafo <code>G</code>, ya que cualquier intento de quitar una arista rompería la conectividad.</li>
        <li><strong>Varios árboles generadores:</strong> En un grafo conectado que no es un árbol, puede haber varios árboles generadores posibles, utilizando diferentes combinaciones de aristas.</li>
    </ul>

    <h2>Aplicaciones de los Árboles Generadores</h2>
    <p>Los árboles generadores son fundamentales en muchas áreas, incluyendo:</p>
    <ul>
        <li><strong>Redes de comunicaciones:</strong> Para diseñar redes eficientes donde cada nodo esté conectado sin redundancia innecesaria.</li>
        <li><strong>Optimización de rutas:</strong> En problemas como el Árbol de Expansión Mínima (<em>Minimum Spanning Tree</em>), donde se busca conectar todos los vértices minimizando el costo total (por ejemplo, la longitud de las aristas).</li>
    </ul>

</body>
</html>
"""

# Mostrar el contenido HTML en Colab
display(HTML(html_content))
```

## 1. ¿Qué es un árbol en teoría de grafos?
Un **árbol** es un grafo no dirigido, conexo y acíclico. Esto significa que no tiene ciclos y hay exactamente **un camino entre cualquier par de vértices**. Los árboles son fundamentales en muchas aplicaciones, como las estructuras de datos y el diseño de redes.

## 2. ¿Qué es un bosque en teoría de grafos?
Un **bosque** es un conjunto de árboles disjuntos. Es un grafo acíclico que puede no estar conectado, es decir, está compuesto por varios árboles separados.

## 3. ¿Qué es un puente en un grafo?
Un **puente** es una arista en un grafo cuya eliminación **desconecta el grafo**. Formalmente, si eliminar una arista aumenta el número de componentes conectados, esa arista es un puente.

## 4. ¿Qué es un árbol generador (spanning tree)?
Un **árbol generador** es un subgrafo de un grafo conexo que contiene todos los vértices del grafo original, no tiene ciclos y es conexo. Su objetivo es conectar todos los vértices minimizando el número de aristas.

## 5. ¿Cuáles son las propiedades de los árboles generadores?
Las principales propiedades de un **árbol generador** son:
- Es **acíclico**.
- Incluye **todos los vértices** del grafo original.
- Tiene exactamente \( V - 1 \) aristas, donde \( V \) es el número de vértices del grafo.

## 6. ¿Cuál es la relación entre árboles y puentes?
En un árbol, **todas las aristas son puentes** porque su eliminación desconectaría el grafo, ya que un árbol no contiene ciclos y tiene el número mínimo de aristas necesarias para conectarlo.

## 7. ¿Qué es el algoritmo de Kruskal?
El **algoritmo de Kruskal** es un algoritmo para encontrar el árbol generador de costo mínimo en un grafo ponderado. Funciona seleccionando las aristas de menor peso y agregándolas al árbol generador siempre que no formen un ciclo.

## 8. ¿Qué es el algoritmo de Prim?
El **algoritmo de Prim** también se utiliza para encontrar el árbol generador de costo mínimo. Comienza con un solo vértice y agrega las aristas más baratas que conectan los vértices en el árbol a los que aún no están incluidos.

## 9. ¿Cuántas aristas tiene un árbol con \( n \) vértices?
Un árbol con \( n \) vértices tiene exactamente **\( n - 1 \) aristas**. Esta es una de las propiedades fundamentales de los árboles.

## 10. ¿Qué es un ciclo en un grafo?
Un **ciclo** es un camino en un grafo que comienza y termina en el mismo vértice sin repetir ninguna arista. Los árboles, por definición, **no contienen ciclos**.

## 11. ¿Cómo se relaciona la teoría de grafos con las redes de comunicaciones?
La teoría de grafos se aplica en **redes de comunicaciones** para optimizar la conectividad. Los árboles generadores mínimos se utilizan para diseñar redes eficientes que conecten todos los nodos sin redundancias innecesarias.

## 12. ¿Qué es un árbol de expansión mínima?
Un **árbol de expansión mínima** es un árbol generador de un grafo ponderado que minimiza la suma de los pesos de las aristas seleccionadas. Es fundamental para problemas de optimización de costos.

## 13. ¿Qué es un subgrafo?
Un **subgrafo** es un conjunto de vértices y aristas que forma parte de un grafo más grande. En el caso de un árbol generador, el subgrafo incluye todos los vértices del grafo original, pero no todos los ciclos.

## 14. ¿Qué es la conectividad en un grafo?
Un grafo es **conexo** si existe un camino entre cualquier par de vértices. Un árbol es siempre un grafo conexo, pero no contiene ciclos.

## 15. ¿Qué son los vértices y aristas en un grafo?
Los **vértices** son los puntos o nodos de un grafo, mientras que las **aristas** son las conexiones entre estos vértices. Un grafo se representa por \( G = (V, E) \), donde \( V \) es el conjunto de vértices y \( E \) el conjunto de aristas.

## 16. ¿Cuál es la importancia de los árboles generadores en la teoría de redes?
Los **árboles generadores** permiten diseñar redes eficientes, como redes eléctricas o de telecomunicaciones, minimizando el costo de conexión entre los nodos sin ciclos redundantes.

## 17. ¿Qué es un ciclo mínimo en un grafo?
Un **ciclo mínimo** es el ciclo de menor longitud posible en un grafo. No es relevante en un árbol, ya que los árboles **no tienen ciclos**.

## 18. ¿Qué es la estructura de un árbol en informática?
En **informática**, un árbol es una estructura de datos jerárquica que simula una estructura de árbol, con un **nodo raíz** y varios subárboles formados por nodos hijos. Es muy útil para búsquedas y ordenación.

## 19. ¿Cómo se determina si un grafo tiene un árbol generador?
Un grafo conexo tiene al menos un **árbol generador**. Si no es conexo, sus componentes también tendrán sus propios árboles generadores, formando un **bosque**.

## 20. ¿Qué aplicaciones tiene el árbol generador en problemas de optimización?
El **árbol generador** se utiliza en problemas de optimización de rutas, donde se busca conectar puntos con el menor costo posible, y en la distribución de redes para minimizar la cantidad de cable o energía necesaria para conectar nodos.
