```python
from IPython.core.display import display, HTML

# Contenido HTML para la Proposición 3.1.3 y BFS
html_content = """
<!DOCTYPE html>
<html lang="es">
<head>
    <meta charset="UTF-8">
    <meta name="viewport" content="width=device-width, initial-scale=1.0">
    <title>Proposición 3.1.3: Árbol Generador</title>
    <style>
        body {
            font-family: Arial, sans-serif;
            line-height: 1.8;
            margin: 20px;
            padding: 20px;
            background-color: #f9f9f9;
        }
        h1, h2, h3 {
            color: #2E8B57;
            margin-top: 20px;
        }
        h1 {
            font-size: 2.5em;
            text-align: center;
        }
        h2 {
            font-size: 1.8em;
        }
        h3 {
            font-size: 1.5em;
        }
        p {
            font-size: 1.2em;
            margin-bottom: 15px;
        }
        ul {
            list-style-type: disc;
            margin-left: 30px;
            margin-bottom: 20px;
        }
        li {
            margin-bottom: 10px;
        }
        code {
            background-color: #f4f4f4;
            padding: 2px 4px;
            font-family: Consolas, "Courier New", monospace;
        }
        pre {
            background-color: #f4f4f4;
            padding: 10px;
            border-radius: 5px;
            font-size: 1em;
            overflow-x: auto;
        }
    </style>
</head>
<body>

    <h1>Proposición 3.1.3: Todo grafo conexo contiene un árbol generador</h1>

    <p>Esta proposición establece que todo grafo <strong>conexo</strong> contiene al menos un <strong>árbol generador</strong> (<em>spanning tree</em>). Es decir, a partir de cualquier grafo conexo (donde cualquier par de vértices está conectado por un camino), es posible encontrar un subgrafo que conecte todos los vértices sin ciclos.</p>

    <h2>Demostración</h2>
    <p>La prueba proporciona una forma sencilla de encontrar un árbol generador. El método consiste en lo siguiente:</p>
    <ul>
        <li><strong>Buscar ciclos</strong> en el grafo.</li>
        <li><strong>Eliminar</strong> una arista de cada ciclo encontrado.</li>
        <li><strong>Repetir</strong> el proceso hasta que no queden ciclos, lo que garantiza que el grafo resultante es acíclico y sigue siendo conexo, formando así un árbol generador.</li>
    </ul>

    <h2>Algoritmo de Búsqueda en Anchura (Breadth-First Search, BFS)</h2>
    <p>Una forma algorítmica de encontrar un árbol generador, directamente relacionada con la búsqueda de ciclos, es mediante el <strong>algoritmo de búsqueda en anchura (BFS)</strong>. Este algoritmo no solo permite encontrar un árbol generador, sino también verificar si un grafo es <strong>bipartito</strong> (divisible en dos conjuntos de vértices, A y B, donde no hay aristas entre vértices del mismo conjunto).</p>

    <h3>Pasos del Algoritmo BFS para Encontrar un Árbol Generador</h3>
    <ol>
        <li><strong>Seleccionar un vértice inicial</strong> <code>v<sub>0</sub></code>: Elegimos cualquier vértice <code>v<sub>0</sub></code> del grafo <code>G</code> como la raíz del árbol generador.</li>
        <li><strong>Construcción del árbol por niveles</strong>:
            <ul>
                <li>Colocamos <code>v<sub>0</sub></code> en el nivel 0 del árbol.</li>
                <li>Identificamos todos los vértices vecinos de <code>v<sub>0</sub></code> (distancia 1), y los agregamos al siguiente nivel del árbol.</li>
                <li>Repetimos el proceso con los vecinos de los vecinos, añadiendo vértices a niveles sucesivos según su distancia mínima a <code>v<sub>0</sub></code>.</li>
            </ul>
        </li>
        <li><strong>Verificación de conectividad</strong>: Si el conjunto de vértices en el árbol es igual al conjunto total de vértices de <code>G</code>, entonces hemos completado el árbol generador.</li>
        <li><strong>Añadir aristas</strong>: En cada paso, se elige un vértice <code>v<sub>i</sub></code> que ya está en el árbol y se agrega una arista hacia su vecino <code>w</code> que aún no está en el árbol.</li>
    </ol>

    <h2>Ejemplo del Diámetro y Radio de un Grafo</h2>
    <p>El <strong>diámetro</strong> de un grafo es la máxima distancia entre dos vértices en un grafo conexo. Si el camino más largo entre dos vértices tiene una longitud de 4, entonces el diámetro es 4.</p>

    <p>El <strong>radio</strong> de un grafo es la mínima distancia <code>r</code> tal que todos los vértices están como máximo a una distancia <code>r</code> de algún vértice central. En un grafo con radio 2, todos los vértices están a lo sumo a una distancia de 2 de un vértice particular.</p>

    <h2>Complejidad del Algoritmo</h2>
    <p>El tiempo de ejecución del algoritmo BFS es <code>O(V(G) + E(G))</code>, donde:</p>
    <ul>
        <li><code>V(G)</code>: Es el número de vértices en el grafo.</li>
        <li><code>E(G)</code>: Es el número de aristas.</li>
    </ul>
    <p>Esto se debe a que cada vértice se visita una vez, y cada arista se explora una vez para asegurar que todos los vértices se hayan conectado.</p>

    <h2>Resumen</h2>
    <ul>
        <li><strong>Proposición 3.1.3:</strong> Todo grafo conexo contiene un árbol generador.</li>
        <li><strong>Búsqueda en Anchura (BFS):</strong> Un algoritmo eficiente para construir un árbol generador y verificar la bipartición de un grafo.</li>
        <li>El algoritmo funciona explorando los vecinos de un vértice inicial en orden de cercanía (distancia), agregando vértices al árbol a medida que se avanza.</li>
    </ul>

</body>
</html>
"""

# Mostrar el contenido HTML en Colab
display(HTML(html_content))
```

```python
import networkx as nx
import matplotlib.pyplot as plt

# Crear el grafo
G = nx.Graph()
edges = [(1, 3), (1, 9), (3, 2), (3, 6), (9, 6), (9, 8), (6, 4), (6, 7), (8, 5)]
G.add_edges_from(edges)

# Aplicar BFS desde el nodo 1
bfs_tree = nx.bfs_tree(G, source=1)

# Definir las posiciones de los nodos para visualización
pos = {
    1: (0, 3),
    3: (-1, 2),
    9: (1, 2),
    2: (-2, 1),
    6: (0, 1),
    8: (2, 1),
    4: (-1, 0),
    7: (1, 0),
    5: (2, 0)
}

# Dibujar el grafo y el árbol BFS
plt.figure(figsize=(8, 6))

# Dibujar el grafo completo en azul claro
nx.draw(G, pos, with_labels=True, node_color='lightgray', edge_color='lightblue', node_size=800, font_size=14, font_weight='bold')

# Resaltar el árbol BFS en negro
nx.draw(bfs_tree, pos, with_labels=True, node_color='lightgray', edge_color='black', node_size=800, font_size=14, font_weight='bold')

plt.title("Árbol de Búsqueda en Amplitud (BFS)")
plt.show()
```

```python
import networkx as nx
import matplotlib.pyplot as plt

# Crear el grafo
G = nx.Graph()
edges = [(1, 3), (1, 9), (3, 2), (3, 6), (9, 6), (9, 8), (6, 4), (6, 7), (8, 5)]
G.add_edges_from(edges)

# Aplicar BFS desde el nodo 1
bfs_tree = nx.bfs_tree(G, source=1)

# Obtener niveles de BFS (distancias desde el nodo raíz)
levels = nx.single_source_shortest_path_length(G, source=1)

# Definir colores para cada nivel
level_colors = {
    0: 'lightgreen',  # Nivel 0 (nodo raíz)
    1: 'lightblue',   # Nivel 1
    2: 'orange',      # Nivel 2
    3: 'red'          # Nivel 3
}

# Crear una lista de colores para los nodos basados en el nivel
node_colors = [level_colors[levels[node]] for node in G.nodes()]

# Definir las posiciones de los nodos para visualización
pos = {
    1: (0, 3),
    3: (-1, 2),
    9: (1, 2),
    2: (-2, 1),
    6: (0, 1),
    8: (2, 1),
    4: (-1, 0),
    7: (1, 0),
    5: (2, 0)
}

# Dibujar el grafo y el árbol BFS
plt.figure(figsize=(8, 6))

# Dibujar el grafo completo en azul claro
nx.draw(G, pos, with_labels=True, node_color=node_colors, edge_color='lightblue', node_size=800, font_size=14, font_weight='bold')

# Resaltar el árbol BFS en negro
nx.draw(bfs_tree, pos, with_labels=True, node_color=node_colors, edge_color='black', node_size=800, font_size=14, font_weight='bold')

plt.title("Árbol de Búsqueda en Amplitud (BFS) con Niveles de Colores")
plt.show()
```

```python
import matplotlib.pyplot as plt
import networkx as nx
import numpy as np

# Crear el Grafo de Petersen
G = nx.petersen_graph()

# Definir la disposición exacta de los nodos basada en la imagen proporcionada
pos = {
    0: (0, 1.5),    # Nodo 1
    1: (-1.4, 0.5), # Nodo 2
    2: (-0.8, -1.2),# Nodo 3
    3: (0.8, -1.2), # Nodo 4
    4: (1.4, 0.5),  # Nodo 5
    5: (0, 0.6),    # Nodo 6
    6: (-0.5, 0.2), # Nodo 7
    7: (-0.4, -0.7),# Nodo 8
    8: (0.4, -0.7), # Nodo 9
    9: (0.5, 0.2)   # Nodo 10
}

# Crear un mapa de colores para las aristas
edge_colors = np.linspace(0, 1, len(G.edges))
edges = list(G.edges)

# Dibujar el grafo con splines (curvas) para las aristas
plt.figure(figsize=(8, 8))

# Dibujar los nodos
nx.draw_networkx_nodes(G, pos, node_color='black', node_size=700)

# Dibujar etiquetas
nx.draw_networkx_labels(G, pos, font_color='white', font_size=12)

# Dibujar las aristas curvas con diferentes colores
nx.draw_networkx_edges(
    G, pos,
    edgelist=edges,
    edge_color=edge_colors,
    edge_cmap=plt.cm.viridis,  # Usamos un mapa de colores (degradado)
    width=2.5,
    connectionstyle='arc3, rad=0.2'  # Las aristas se dibujan con curvas
)

# Mostrar el grafo
plt.title('Grafo de Petersen con aristas dinámicas')
plt.axis('off')  # Ocultar los ejes
plt.show()
```

## 1. ¿Qué establece la Proposición 3.1.3 sobre los grafos conexos y los árboles generadores?
La **Proposición 3.1.3** establece que todo grafo conexo contiene al menos un **árbol generador** (spanning tree). Esto significa que es posible encontrar un subgrafo acíclico que conecte todos los vértices del grafo original, conservando su conectividad.

## 2. ¿Cómo se define un árbol generador en un grafo?
Un **árbol generador** es un subgrafo de un grafo conexo que incluye todos los vértices del grafo original y es **acíclico** (sin ciclos). Además, tiene exactamente \( V - 1 \) aristas, donde \( V \) es el número de vértices del grafo.

## 3. ¿Cuál es el proceso para encontrar un árbol generador en un grafo conexo?
El proceso para encontrar un árbol generador implica identificar y eliminar ciclos del grafo. A medida que se detectan ciclos, se eliminan aristas hasta que el grafo queda sin ciclos y mantiene su conectividad, transformándose en un **árbol generador**.

## 4. ¿Qué rol juegan los ciclos en la construcción de un árbol generador?
Los **ciclos** se eliminan durante la construcción de un árbol generador. En un árbol generador no puede haber ciclos, ya que estos introducirían redundancia en la conectividad. El proceso implica encontrar ciclos y eliminar una arista de cada ciclo hasta que el grafo se vuelva acíclico.

## 5. ¿Qué es el algoritmo de Búsqueda en Anchura (BFS) y cómo se relaciona con los árboles generadores?
El **algoritmo de Búsqueda en Anchura (BFS)** es un algoritmo que recorre un grafo **nivel por nivel**, comenzando desde un vértice inicial y explorando todos sus vecinos antes de pasar al siguiente nivel de vértices. Este algoritmo se puede usar para construir un **árbol generador**, conectando todos los vértices sin crear ciclos.

## 6. ¿Cuáles son los pasos del algoritmo BFS para encontrar un árbol generador?
Los pasos del algoritmo BFS para encontrar un árbol generador son:
1. Seleccionar un vértice inicial \( v_0 \).
2. Construir el árbol **nivel por nivel**, agregando los vecinos de cada vértice al nivel correspondiente.
3. Verificar la **conectividad** del grafo.
4. Añadir las aristas necesarias sin formar ciclos, completando así el árbol generador.

## 7. ¿Cómo se selecciona el vértice inicial en el algoritmo BFS y por qué es importante?
En el algoritmo BFS, se selecciona un vértice inicial \( v_0 \) de manera arbitraria, y este se utiliza como la **raíz del árbol**. La selección del vértice inicial es importante porque determina el punto de partida para la exploración de niveles en el árbol BFS.

## 8. ¿Qué significa construir un árbol BFS por niveles?
Construir un árbol BFS por niveles implica organizar los vértices en función de su **distancia** al vértice raíz \( v_0 \). Los vértices directamente conectados con \( v_0 \) forman el **nivel 1**, los vértices conectados con los vértices del nivel 1 forman el **nivel 2**, y así sucesivamente.

## 9. ¿Cómo se verifica la conectividad de un grafo utilizando el algoritmo BFS?
La conectividad se verifica en el algoritmo BFS al comprobar que **todos los vértices** del grafo original han sido visitados y añadidos al árbol BFS. Si todos los vértices están presentes en el árbol generador, el grafo es conexo.

## 10. ¿Qué función cumplen las aristas en la construcción del árbol generador con BFS?
Las aristas en el algoritmo BFS conectan los vértices en cada nivel del árbol, pero se seleccionan cuidadosamente para **evitar la creación de ciclos**. Solo se añaden aristas entre vértices que aún no están conectados en el árbol generador.

## 11. ¿Cómo se define el diámetro de un grafo?
El **diámetro de un grafo** es la distancia más larga entre dos vértices del grafo. Es decir, es la longitud del **camino más largo** entre cualquier par de vértices en el grafo conexo.

## 12. ¿Qué es el radio de un grafo y cómo se calcula?
El **radio de un grafo** es la mínima distancia \( r \) tal que todos los vértices del grafo están como máximo a esa distancia de un **vértice central**. Se calcula encontrando el vértice con la menor distancia máxima a todos los demás vértices.

## 13. ¿Cuál es la complejidad temporal del algoritmo BFS y cómo se calcula?
La **complejidad temporal** del algoritmo BFS es \( O(V(G) + E(G)) \), donde \( V(G) \) es el número de vértices y \( E(G) \) es el número de aristas en el grafo. Esta complejidad se debe a que el algoritmo visita cada vértice y explora cada arista una vez.

## 14. ¿Qué propiedades del grafo se pueden determinar utilizando BFS, además del árbol generador?
Además de construir un árbol generador, el algoritmo BFS puede determinar si un grafo es **bipartito** (si se puede dividir en dos conjuntos sin aristas internas) y puede calcular otras propiedades como el **diámetro** y el **radio** del grafo.

## 15. ¿Cómo se utiliza el algoritmo BFS para organizar un grafo en niveles?
El algoritmo BFS organiza un grafo en niveles comenzando desde un **vértice raíz**. Los vecinos inmediatos del vértice raíz forman el **primer nivel**, los vecinos de estos forman el **segundo nivel**, y así sucesivamente, creando una jerarquía basada en distancias.

## 16. ¿Qué es el grafo de Petersen y cuáles son sus características?
El **grafo de Petersen** es un grafo no planar con 10 vértices y 15 aristas. Tiene varias propiedades interesantes, como que **no es bipartito**, tiene un **diámetro de 2**, y una alta **conectividad**, lo que lo convierte en un grafo ampliamente estudiado en teoría de grafos.

## 17. ¿Cómo se aplica el algoritmo BFS al grafo de Petersen para encontrar un árbol generador?
En el grafo de Petersen, el algoritmo BFS se aplica seleccionando un vértice raíz (por ejemplo, el vértice 1), y luego explorando sus vecinos nivel por nivel, añadiendo aristas entre los vértices que no crean ciclos, hasta completar el árbol generador.

## 18. ¿Cómo se organizan los vértices del grafo de Petersen en niveles mediante BFS?
Los vértices del grafo de Petersen se organizan en niveles basados en su **distancia al vértice raíz**. Los vecinos del vértice raíz forman el **nivel 1**, los vecinos de los vértices del nivel 1 forman el **nivel 2**, y así sucesivamente hasta incluir todos los vértices.

## 19. ¿Qué significa que el grafo de Petersen tiene un diámetro de 2 y un radio de 2?
El **diámetro de 2** significa que la distancia más larga entre dos vértices del grafo es de 2 aristas. El **radio de 2** indica que todos los vértices están a una distancia máxima de 2 aristas desde un **vértice central**.

## 20. ¿Cómo se visualizan los niveles y aristas en un árbol BFS construido a partir de un grafo?
Los niveles y aristas en un árbol BFS se visualizan colocando el **vértice raíz** en el nivel 0 y organizando los demás vértices en niveles según su **distancia al vértice raíz**. Las aristas conectan los vértices entre niveles, resaltando las conexiones sin crear ciclos.
