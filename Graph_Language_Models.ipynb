```python
from IPython.display import display, HTML

# Contenido HTML
html_content = """
<!DOCTYPE html>
<html lang="es">
<head>
    <meta charset="UTF-8">
    <meta name="viewport" content="width=device-width, initial-scale=1.0">
    <title>Modelos de Generación Basados en Grafos y Texto</title>
    <style>
        body {
            font-family: Arial, sans-serif;
            line-height: 1.8;
            margin: 20px;
            padding: 20px;
            background-color: #f9f9f9;
        }
        h1, h2 {
            color: #333;
            font-size: 2em;
            margin-bottom: 20px;
        }
        p {
            font-size: 1.2em;
            margin-bottom: 15px;
        }
        ul {
            list-style-type: square;
            margin-left: 30px;
            margin-bottom: 20px;
        }
        li {
            margin-bottom: 10px;
        }
        a {
            color: #0066cc;
            text-decoration: none;
            font-weight: bold;
        }
        a:hover {
            text-decoration: underline;
        }
        .button-link {
            background-color: #0066cc;
            color: white;
            padding: 10px 15px;
            text-align: center;
            display: inline-block;
            border-radius: 5px;
            margin-top: 20px;
            text-decoration: none;
            font-size: 1.2em;
        }
        .button-link:hover {
            background-color: #004999;
        }
    </style>
</head>
<body>

    <h1>Definición y Explicación Teórica de los Modelos de Generación Basados en Grafos y Texto</h1>

    <p>Los modelos de generación basados en grafos y texto combinan la capacidad de razonar sobre datos estructurados, como los grafos, con la comprensión del lenguaje natural. Este enfoque permite que un modelo genere texto o haga predicciones que tengan en cuenta tanto las relaciones explícitas entre entidades, que son modeladas mediante grafos, como el contexto proporcionado por el texto. Estos modelos son una extensión de las redes neuronales profundas y los transformers, como el modelo T5 (Text-to-Text Transfer Transformer), que procesan texto y grafos conjuntamente.</p>

    <h2>¿Qué es un Grafo?</h2>
    <p>Un grafo es una estructura matemática que representa objetos (nodos) y las relaciones entre ellos (aristas o conexiones). Un grafo está compuesto por:</p>
    <ul>
        <li><strong>Nodos:</strong> Representan entidades o elementos (por ejemplo, personas, objetos, conceptos).</li>
        <li><strong>Aristas:</strong> Representan las relaciones entre esos nodos (por ejemplo, una relación de "es un" entre un "perro" y un "animal").</li>
    </ul>

    <p><strong>Ejemplo de un grafo:</strong></p>
    <ul>
        <li><strong>Nodo 1:</strong> "perro"</li>
        <li><strong>Nodo 2:</strong> "animal"</li>
        <li><strong>Arista:</strong> "es un", que conecta el nodo "perro" con "animal".</li>
    </ul>

    <h2>Modelos Basados en Grafos y Texto</h2>
    <p>Los modelos basados en grafos y texto permiten razonar sobre datos que tienen una estructura relacional (grafos) y texto no estructurado al mismo tiempo. Estos modelos usan técnicas avanzadas de aprendizaje automático para procesar ambos tipos de datos, combinando información relacional con contexto lingüístico.</p>

    <h2>Arquitectura Típica</h2>
    <p>La arquitectura de estos modelos suele basarse en transformers, como el T5 (Text-to-Text Transfer Transformer), que tiene la capacidad de generar secuencias de texto condicionalmente, basándose en entradas tanto de texto como de grafos. Los pasos principales en esta arquitectura son:</p>

    <ul>
        <li><strong>Entrada:</strong>
            <ul>
                <li><strong>Grafo:</strong> El grafo representa relaciones entre entidades (por ejemplo, "el perro es un animal").</li>
                <li><strong>Texto:</strong> El texto puede proporcionar contexto adicional al grafo (por ejemplo, "el perro persiguió al gato").</li>
            </ul>
        </li>
        <li><strong>Codificación:</strong>
            <ul>
                <li>Los nodos y las aristas del grafo se transforman en representaciones numéricas (embeddings) mediante un encoder.</li>
                <li>El texto también se transforma en embeddings, utilizando un tokenizador que convierte las palabras en secuencias de tokens.</li>
            </ul>
        </li>
        <li><strong>Fusión de la Información:</strong>
            <ul>
                <li>Las representaciones de los grafos y el texto se combinan para crear una representación conjunta.</li>
                <li>Esta combinación permite que el modelo entienda tanto las relaciones explícitas entre las entidades como el contexto lingüístico en el que esas entidades se encuentran.</li>
            </ul>
        </li>
        <li><strong>Generación:</strong>
            <ul>
                <li>A partir de las representaciones conjuntas, un decodificador genera texto nuevo. Este texto puede ser un resumen, una predicción de relaciones, o una respuesta a una pregunta basada en el grafo y el texto proporcionado.</li>
            </ul>
        </li>
    </ul>

    <h2>Aplicaciones</h2>
    <p>Los modelos que combinan grafos y texto tienen una amplia gama de aplicaciones en campos donde es esencial razonar sobre entidades y sus relaciones dentro de un contexto más amplio proporcionado por el lenguaje natural.</p>

    <ul>
        <li><strong>1. Sistemas de Respuesta a Preguntas (QA) Basados en Conocimiento:</strong> En estos sistemas, los grafos de conocimiento (Knowledge Graphs) representan relaciones entre entidades (por ejemplo, personas, lugares, eventos), y el modelo puede generar respuestas a preguntas complejas utilizando tanto el grafo como el texto relacionado.
            <br><em>Ejemplo:</em> Preguntar "¿Quién es el presidente de Francia?" requiere que el modelo consulte un grafo que conecte a la persona con el país y también utilice información textual para proporcionar la respuesta.</li>
        <li><strong>2. Generación de Texto Basada en Datos Estructurados:</strong> Los modelos pueden generar descripciones de entidades o eventos basados en las relaciones entre los nodos de un grafo.
            <br><em>Ejemplo:</em> Generar automáticamente un resumen de un evento deportivo, donde el grafo contiene la información estructurada del juego (jugadores, puntuaciones, tiempos) y el modelo lo transforma en un texto narrativo.</li>
        <li><strong>3. Predicción de Relaciones en Ciencia y Medicina:</strong> En aplicaciones científicas, los grafos pueden representar interacciones entre moléculas o proteínas, y el texto científico proporciona contexto adicional.
            <br><em>Ejemplo:</em> Un modelo podría predecir interacciones entre medicamentos utilizando un grafo que represente las relaciones entre compuestos químicos y artículos científicos que describan experimentos.</li>
        <li><strong>4. Recomendaciones en Redes Sociales:</strong> Los grafos que representan conexiones entre usuarios, junto con el texto de publicaciones, se pueden utilizar para generar recomendaciones personalizadas, como sugerencias de amigos o contenido.
            <br><em>Ejemplo:</em> Un modelo que combine el grafo de conexiones entre usuarios de una red social con las publicaciones que estos han hecho para recomendar nuevos amigos o grupos.</li>
        <li><strong>5. Procesamiento de Datos Financieros:</strong> Los grafos pueden representar las relaciones entre diferentes entidades financieras, y el texto puede ser artículos de noticias o informes financieros.
            <br><em>Ejemplo:</em> Un modelo podría predecir el impacto de un evento económico en una red de empresas basándose en un grafo de relaciones comerciales y en informes de mercado.</li>
    </ul>

    <h2>Ventajas de los Modelos Basados en Grafos y Texto</h2>
    <ul>
        <li><strong>Combinación de Información Estructurada y No Estructurada:</strong> Estos modelos pueden procesar datos estructurados (como grafos de conocimiento) y no estructurados (como texto en lenguaje natural) simultáneamente, lo que los hace muy versátiles.</li>
        <li><strong>Capacidad de Razonamiento:</strong> Al modelar explícitamente las relaciones entre entidades a través de grafos, estos modelos pueden realizar inferencias más complejas y precisas.</li>
        <li><strong>Aplicaciones Multidominio:</strong> Desde redes sociales hasta biomedicina, estos modelos pueden ser aplicados en múltiples áreas donde se requiera análisis de relaciones complejas entre entidades y comprensión de texto.</li>
    </ul>

    <h2>Material Elaborado por Sergio Gevatschnaider</h2>

    <p>Aquí tienes algunos enlaces útiles:</p>
    <p><a href="https://huggingface.co/models?other=glm-t5" target="_blank" class="button-link">Ver modelos GLM-T5 en Hugging Face</a></p>
    <p><a href="https://www.arxiv.org/pdf/2408.10700" target="_blank" class="button-link">Consultar artículo en ArXiv</a></p>
    <p><a href="https://aclanthology.org/2024.acl-long.245.pdf" target="_blank" class="button-link">Consultar artículo en ACL Anthology</a></p>

</body>
</html>
"""

# Mostrar el contenido HTML en Colab
display(HTML(html_content))
```

```python
!pip install transformers

from transformers import AutoTokenizer, AutoModel

# Nombre del modelo
modelcard = 'plenz/GLM-t5-large'

print('Cargando el modelo y el tokenizador...')
model = AutoModel.from_pretrained(modelcard, trust_remote_code=True, revision='main')
tokenizer = AutoTokenizer.from_pretrained(modelcard)

# Definimos grafos y textos de ejemplo
print('Definiendo entradas de ejemplo (grafos y texto)...')
graph_1 = [
    ('black poodle', 'is a', 'dog'),
    ('dog', 'is a', 'animal'),
    ('cat', 'is a', 'animal')
]
text_1 = 'The dog chased the cat.'

graph_2 = [
    ('dog', 'is a', 'animal'),
    ('dog', 'has', 'tail'),
    ('dog', 'has', 'fur'),
    ('fish', 'is a', 'animal'),
    ('fish', 'has', 'scales')
]
text_2 = None  # Solo grafo para esta instancia

# Preparar las entradas del modelo
print('Preparando entradas del modelo...')
how = 'global'  # Puede ser 'global' o 'local'
data_1 = model.data_processor.encode_graph(tokenizer=tokenizer, g=graph_1, text=text_1, how=how)
data_2 = model.data_processor.encode_graph(tokenizer=tokenizer, g=graph_2, text=text_2, how=how)
datas = [data_1, data_2]
model_inputs = model.data_processor.to_batch(data_instances=datas, tokenizer=tokenizer, max_seq_len=None, device='cpu')

# Computar los embeddings de los tokens
print('Computando codificaciones de tokens...')
outputs = model(**model_inputs)

# Ver los resultados: forma de los embeddings
print('Secuencia de tokens (batch_size, max_seq_len, embedding_dim):', outputs.last_hidden_state.shape)

# Obtener el embedding específico de 'black poodle'
print('Embedding de "black poodle" en la primera instancia:')
black_poodle_embedding = model.data_processor.get_embedding(sequence_embedding=outputs.last_hidden_state[0], indices=data_1.indices, concept='black poodle', embedding_aggregation='seq')
print('Shape del embedding:', black_poodle_embedding.shape)
```

```python
import matplotlib.pyplot as plt
import networkx as nx

# Función para crear un diagrama similar al que mencionas
def visualize_glm_flow():
    plt.figure(figsize=(8, 6))

    # Crear un grafo para representar el flujo
    G = nx.DiGraph()

    # Añadir los nodos principales: Graph Transformer, Language Model, Graph Language Model
    G.add_node("Graph Transformer", pos=(0, 1))
    G.add_node("Language Model", pos=(1, 1))
    G.add_node("Graph Language Model", pos=(0.5, 0.5))

    # Añadir los grafos de entrada y salida
    G.add_node("Don Quijote", pos=(-0.5, 1.5))
    G.add_node("Molinos de viento", pos=(-0.5, 0.5))
    G.add_node("Texto narrativo", pos=(1.5, 1.5))

    G.add_node("Texto Generado", pos=(0.5, -0.2))

    # Conectar los nodos
    G.add_edges_from([("Graph Transformer", "Graph Language Model"),
                      ("Language Model", "Graph Language Model"),
                      ("Graph Language Model", "Texto Generado")])

    G.add_edges_from([("Don Quijote", "Graph Transformer"),
                      ("Molinos de viento", "Graph Transformer"),
                      ("Texto narrativo", "Language Model")])

    # Posiciones de los nodos
    pos = nx.get_node_attributes(G, 'pos')

    # Dibujar el grafo
    nx.draw(G, pos, with_labels=True, node_size=4000, node_color='lightblue', font_size=10, font_weight='bold', arrows=True, arrowsize=20)

    # Añadir etiquetas a las conexiones
    edge_labels = {("Graph Transformer", "Graph Language Model"): "Integración",
                   ("Language Model", "Graph Language Model"): "Parámetros",
                   ("Graph Language Model", "Texto Generado"): "Texto"}
    nx.draw_networkx_edge_labels(G, pos, edge_labels=edge_labels, font_color='black', font_size=12)

    plt.title("Flujo de Generación de Texto basado en Don Quijote", fontsize=16)
    plt.show()

# Llamar a la función para mostrar la visualización
visualize_glm_flow()
```

```python
import matplotlib.pyplot as plt
import networkx as nx
import imageio.v2 as imageio  # Usar imageio.v2 para evitar la advertencia
import os
import numpy as np
from google.colab import drive

# Montar Google Drive para guardar el archivo
drive.mount('/content/drive')

# Crear la carpeta "GIF" si no existe
gif_dir = '/content/drive/MyDrive/GIF'
if not os.path.exists(gif_dir):
    os.makedirs(gif_dir)

# Función para interpolar la posición de las flechas
def interpolate_positions(pos1, pos2, fraction):
    x1, y1 = pos1
    x2, y2 = pos2
    return (x1 + (x2 - x1) * fraction, y1 + (y2 - y1) * fraction)

# Función para crear una imagen del diagrama en cada paso
def create_glm_image(step, save_path, edge_progress=None):
    plt.figure(figsize=(8, 6))
    G = nx.DiGraph()

    # Añadir nodos principales según el paso
    if step >= 1:
        G.add_node("Graph Transformer", pos=(0, 1))
        G.add_node("Don Quijote", pos=(-0.5, 1.5))
        G.add_node("Molinos de viento", pos=(-0.5, 0.5))

    if step >= 2:
        G.add_node("Language Model", pos=(1, 1))
        G.add_node("Texto narrativo", pos=(1.5, 1.5))

    if step >= 3:
        G.add_node("Graph Language Model", pos=(0.5, 0.5))

    if step >= 4:
        G.add_node("Texto Generado", pos=(0.5, -0.2))

    # Posiciones de los nodos
    pos = nx.get_node_attributes(G, 'pos')

    # Dibujar nodos
    nx.draw(G, pos, with_labels=True, node_size=4000, node_color='lightblue', font_size=10, font_weight='bold')

    # Conectar nodos según el paso
    edges = []
    if step >= 1:
        edges.append(("Don Quijote", "Graph Transformer"))
        edges.append(("Molinos de viento", "Graph Transformer"))

    if step >= 2:
        edges.append(("Texto narrativo", "Language Model"))

    if step >= 3:
        edges.append(("Graph Transformer", "Graph Language Model"))
        edges.append(("Language Model", "Graph Language Model"))

    if step >= 4:
        edges.append(("Graph Language Model", "Texto Generado"))

    # Dibujar las flechas de forma progresiva
    if edge_progress:
        for (source, target), fraction in zip(edges, edge_progress):
            pos_source = pos[source]
            pos_target = pos[target]
            intermediate_pos = interpolate_positions(pos_source, pos_target, fraction)
            # Dibuja la flecha de color negro completa
            nx.draw_networkx_edges(G, pos, edgelist=[(source, target)], arrowstyle='-|>', arrowsize=20,
                                   edge_color='black', min_source_margin=15, min_target_margin=15)
            # Dibuja la parte roja que muestra el progreso
            nx.draw_networkx_edges(G, {source: pos_source, target: intermediate_pos}, edgelist=[(source, target)],
                                   arrowstyle='-|>', arrowsize=20, edge_color='red')

    # Etiquetas de las conexiones
    if step >= 3 and all(fraction == 1 for fraction in edge_progress):
        edge_labels = {("Graph Transformer", "Graph Language Model"): "Integración",
                       ("Language Model", "Graph Language Model"): "Parámetros"}
        nx.draw_networkx_edge_labels(G, pos, edge_labels=edge_labels, font_color='black', font_size=12)

    if step >= 4 and all(fraction == 1 for fraction in edge_progress):
        edge_labels = {("Graph Language Model", "Texto Generado"): "Texto"}
        nx.draw_networkx_edge_labels(G, pos, edge_labels=edge_labels, font_color='black', font_size=12)

    plt.title(f"Paso {step} en el Flujo de Datos", fontsize=16)

    # Guardar la imagen
    plt.savefig(save_path)
    plt.close()

# Función para crear el GIF dinámico continuo y guardarlo en Google Drive
def create_continuous_glm_gif():
    # Crear una carpeta temporal para las imágenes
    if not os.path.exists('glm_frames'):
        os.makedirs('glm_frames')

    # Generar imágenes para cada paso del proceso
    total_steps = 4  # Número de pasos en el flujo
    frame_count_per_step = 30  # Número de frames por paso para suavizar la animación
    images = []

    for frame in range(total_steps * frame_count_per_step):
        # Calcular el paso actual y la fracción completada dentro de ese paso
        step = frame // frame_count_per_step + 1
        frame_fraction = (frame % frame_count_per_step) / frame_count_per_step

        # Progresión de las aristas, interpolamos aquí
        edge_progress = [min(frame_fraction, 1)] * len([
            ("Don Quijote", "Graph Transformer"),
            ("Molinos de viento", "Graph Transformer"),
            ("Texto narrativo", "Language Model"),
            ("Graph Transformer", "Graph Language Model"),
            ("Language Model", "Graph Language Model"),
            ("Graph Language Model", "Texto Generado")
        ][:step])

        image_path = f'glm_frames/continuous_step_{step}_frame_{frame}.png'
        create_glm_image(step, image_path, edge_progress=edge_progress)
        images.append(imageio.imread(image_path))

    # Guardar el GIF en la carpeta "GIF" de Google Drive
    gif_path = f'{gif_dir}/glm_process_continuous_fixed.gif'
    imageio.mimsave(gif_path, images, duration=0.1)

    # Limpiar las imágenes temporales solo después de crear el GIF
    for frame in range(total_steps * frame_count_per_step):
        step = frame // frame_count_per_step + 1
        os.remove(f'glm_frames/continuous_step_{step}_frame_{frame}.png')

    # Confirmación
    print(f"El GIF continuo se ha guardado en tu Google Drive en la carpeta GIF como 'glm_process_continuous_fixed.gif'.")

# Crear el GIF continuo y guardarlo en la carpeta GIF en Google Drive
create_continuous_glm_gif()

# Mostrar el GIF en Google Colab
from IPython.display import Image
Image(filename=f'{gif_dir}/glm_process_continuous_fixed.gif')
```
