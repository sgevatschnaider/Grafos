```python
import networkx as nx
import matplotlib.pyplot as plt

# Función para mostrar el grafo con sus aristas resaltadas
def draw_graph_with_highlighted_edges(G, highlighted_edges, title):
    pos = nx.spring_layout(G)  # Posición de los nodos para visualización

    # Dibuja el grafo base
    nx.draw(G, pos, with_labels=True, node_color='lightblue', node_size=600, font_weight='bold', font_size=10)

    # Dibuja las aristas resaltadas
    nx.draw_networkx_edges(G, pos, edgelist=highlighted_edges, edge_color='r', width=3)

    # Título del grafo
    plt.title(title)
    plt.show()

# 1. Grafo Euleriano pero no Hamiltoniano
# Un ciclo de 4 nodos con un vértice extra conectado a uno de ellos
G_eulerian = nx.Graph()
edges_eulerian = [(1, 2), (2, 3), (3, 4), (4, 1), (1, 5)]
G_eulerian.add_edges_from(edges_eulerian)

# Este grafo es euleriano pero no hamiltoniano
eulerian_circuit = list(nx.eulerian_circuit(G_eulerian, source=1))

# Convertir el circuito de euler en lista de aristas para visualización
highlighted_edges_eulerian = [(u, v) for u, v in eulerian_circuit]

draw_graph_with_highlighted_edges(G_eulerian, highlighted_edges_eulerian, "Grafo Euleriano (No Hamiltoniano)")

# 2. Grafo Hamiltoniano pero no Euleriano
# Un ciclo de 5 nodos, donde cada nodo tiene un grado impar
G_hamiltonian = nx.Graph()
edges_hamiltonian = [(1, 2), (2, 3), (3, 4), (4, 5), (5, 1)]
G_hamiltonian.add_edges_from(edges_hamiltonian)

# Para el ciclo hamiltoniano, simplemente usamos todas las aristas del ciclo
hamiltonian_cycle = edges_hamiltonian  # Todas las aristas son parte del ciclo hamiltoniano

draw_graph_with_highlighted_edges(G_hamiltonian, hamiltonian_cycle, "Grafo Hamiltoniano (No Euleriano)")
```

```python
from typing import List, Dict, Set
from collections import defaultdict

class Graph:
    def __init__(self, edges: List[tuple[int, int]]):
        self.graph: Dict[int, Set[int]] = defaultdict(set)
        self.visited = set()
        self.total_nodes = 0
        self.make_graph(edges)

    def make_graph(self, edges: List[tuple[int, int]]) -> None:
        for u, v in edges:
            self.graph[u].add(v)
            self.graph[v].add(u)
        self.total_nodes = len(self.graph)

    def get_hamiltonian_path(self, start: int) -> List[List[int]]:
        """Obtiene todos los caminos Hamiltonianos desde un nodo inicial"""
        self.visited = set()  # Resetea nodos visitados
        return self._find_paths(start, [])

    def _find_paths(self, node: int, path: List[int]) -> List[List[int]]:
        path.append(node)
        self.visited.add(node)

        if len(self.visited) == self.total_nodes:
            # Si todos los nodos han sido visitados, retorna este camino como válido
            result = [path[:]]  # Hace una copia del camino actual
        else:
            result = []
            for neighbor in self.graph[node]:
                if neighbor not in self.visited:
                    result.extend(self._find_paths(neighbor, path))

        path.pop()  # Retira el nodo actual para explorar otros caminos
        self.visited.remove(node)
        return result

# Ejemplo de uso
if __name__ == '__main__':
    edges = [
        (1, 2), (2, 3), (3, 4), (4, 1), (2, 4), (1, 3)
    ]
    graph = Graph(edges)
    hamiltonian_paths = graph.get_hamiltonian_path(start=1)

    for path in hamiltonian_paths:
        print(" -> ".join(map(str, path)))
```

```python
# Importar las bibliotecas necesarias
import networkx as nx
import matplotlib.pyplot as plt

# Configurar estilos de visualización
plt.style.use('seaborn-darkgrid')

# Crear un grafo hamiltoniano: Ciclo de 6 vértices
G_hamiltonian = nx.cycle_graph(6)

# Crear un grafo no hamiltoniano: Grafo Estrella con 5 vértices
G_non_hamiltonian = nx.star_graph(4)  # Nodo central 0 conectado a nodos 1,2,3,4

# Asignar posiciones para una visualización clara
pos_hamiltonian = nx.circular_layout(G_hamiltonian)
pos_non_hamiltonian = nx.spring_layout(G_non_hamiltonian)

# Dibujar los grafos
plt.figure(figsize=(16, 8))

# Grafo Hamiltoniano
plt.subplot(1, 2, 1)
nx.draw(G_hamiltonian, pos_hamiltonian, with_labels=True,
        node_color='#85C1E9', edge_color='#2980B9',
        node_size=700, font_weight='bold', linewidths=2)
plt.title("Grafo Hamiltoniano", fontsize=18)

# Grafo No Hamiltoniano
plt.subplot(1, 2, 2)
nx.draw(G_non_hamiltonian, pos_non_hamiltonian, with_labels=True,
        node_color='#F1948A', edge_color='#E74C3C',
        node_size=700, font_weight='bold', linewidths=2)
plt.title("Grafo No Hamiltoniano", fontsize=18)

plt.tight_layout()
plt.show()
```

1.	¿Qué es un ciclo hamiltoniano?

Un ciclo hamiltoniano es un ciclo en un grafo que visita exactamente una vez cada uno de los vértices del grafo y regresa al vértice de inicio. Es decir, es un recorrido que pasa por todos los vértices del grafo sin repetir ninguno, y luego vuelve al punto de partida. Este tipo de ciclos son importantes en aplicaciones como la optimización de rutas en redes o problemas como el problema del viajante.
2.	¿Qué es un camino hamiltoniano?

Un camino hamiltoniano es un camino en un grafo que visita exactamente una vez cada vértice, pero no está obligado a regresar al vértice de inicio. A diferencia del ciclo hamiltoniano, no forma un ciclo cerrado, aunque sigue recorriendo todos los vértices una sola vez. Es útil en problemas donde se requiere pasar por todos los puntos sin necesidad de volver al inicio, como en ciertas redes de transporte o logística.

3.	¿Qué establece el Teorema de Dirac sobre los grafos hamiltonianos?

El Teorema de Dirac proporciona una condición suficiente para la existencia de un ciclo hamiltoniano en un grafo. El teorema establece lo siguiente:

Dado un grafo $G$ con $n$ vértices (donde $n \geq 3$), si el grado mínimo de cada vértice es al menos $\frac{n}{2}$ (es decir, cada vértice está conectado con al menos la mitad de los otros vértices), entonces el grafo contiene un ciclo hamiltoniano.

Esta no es una condición necesaria, pero sí una guía útil para detectar grafos que podrían ser hamiltonianos.


4.	¿Cuál es la diferencia entre grafos eulerianos y hamiltonianos?
Los grafos eulerianos y hamiltonianos son conceptos similares pero se refieren a recorridos diferentes en un grafo. Un grafo euleriano contiene un ciclo euleriano, que es un ciclo que pasa por todas las aristas exactamente una vez. Por otro lado, un grafo hamiltoniano contiene un ciclo hamiltoniano, que pasa por todos los vértices una sola vez. La principal diferencia es que los grafos eulerianos están enfocados en las aristas, mientras que los hamiltonianos se enfocan en los vértices.


5.	¿Qué es un grafo euleriano?
Un grafo euleriano es un grafo en el cual existe un ciclo que recorre todas las aristas exactamente una vez y regresa al punto de inicio. Para que un grafo sea euleriano, es necesario que todos sus vértices tengan un grado par (es decir, que el número de aristas conectadas a cada vértice sea par). Este tipo de grafo es fundamental en problemas de diseño de circuitos, recorridos postales o rutas de entrega, donde se necesita visitar todas las conexiones sin repetir ninguna.

6.	¿Qué es un ciclo euleriano?


Un ciclo euleriano es un ciclo en un grafo que recorre todas las aristas del grafo exactamente una vez y regresa al vértice de inicio. Es una propiedad crucial en los grafos eulerianos. Un ciclo euleriano no necesariamente visita todos los vértices del grafo, solo se asegura de cubrir todas las aristas. Este tipo de ciclo fue descubierto por Leonhard Euler y tiene aplicaciones en planificación de rutas que deben cubrir todas las conexiones de una red.

7.	¿Qué condición suficiente proporciona el Teorema de Ore para los grafos hamiltonianos?

8.	¿Qué diferencia existe entre un ciclo y un camino en un grafo?
Un camino en un grafo es una secuencia de vértices conectados por aristas en la que se permite que el recorrido termine en un vértice diferente al de inicio. En contraste, un ciclo es un camino que empieza y termina en el mismo vértice, formando un bucle cerrado. En otras palabras, un ciclo es un camino que regresa a su punto de partida. Ambos conceptos son fundamentales para analizar estructuras en redes y procesos en grafos.


9.	¿Qué es un grafo?

Un grafo es una estructura matemática utilizada para modelar relaciones entre objetos. Está compuesto por vértices (también llamados nodos) que representan los objetos, y aristas (o enlaces) que representan las conexiones entre ellos. Los grafos pueden ser dirigidos o no dirigidos, dependiendo de si las conexiones tienen una dirección específica. Los grafos son ampliamente usados en la informática, redes sociales, análisis de datos, biología, y muchas otras áreas.

10.	¿Qué es un grafo completo?

Un grafo completo es un grafo en el que existe una arista entre cada par de vértices. Los grafos completos son una representación ideal de redes totalmente conectadas y se usan en teoría de juegos y optimización.
