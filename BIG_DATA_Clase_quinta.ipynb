```python
# Importar las librerías necesarias
from IPython.core.display import display, HTML

# Contenido HTML que se mostrará en Colab
html_content = """
<!DOCTYPE html>
<html lang="es">
<head>
    <meta charset="UTF-8">
    <meta name="viewport" content="width=device-width, initial-scale=1.0">
    <title>Handshaking Lemma</title>
    <script type="text/javascript" async
      src="https://cdnjs.cloudflare.com/ajax/libs/mathjax/2.7.7/MathJax.js?config=TeX-MML-AM_CHTML">
    </script>
    <style>
        body {
            font-family: Arial, sans-serif;
            line-height: 1.6;
            background-color: #f0f8ff;
            margin: 0;
            padding: 20px;
        }
        h1, h2 {
            color: #2c3e50;
        }
        p {
            background: #ffffff;
            padding: 10px;
            border-radius: 5px;
            box-shadow: 0 0 10px rgba(0, 0, 0, 0.1);
            margin-bottom: 10px;
        }
        ul {
            list-style-type: disc;
            background: #ffffff;
            padding: 10px;
            border-radius: 5px;
            box-shadow: 0 0 10px rgba(0, 0, 0, 0.1);
        }
        li {
            margin-bottom: 8px;
        }
        .highlight {
            background-color: #ffeb3b;
            padding: 2px 5px;
            border-radius: 3px;
            font-weight: bold;
        }
        .math {
            font-family: 'Cambria Math', 'Times New Roman', serif;
            font-size: 1.2em;
            text-align: center;
            margin: 20px 0;
        }
    </style>
</head>
<body>
    <h1>Handshaking Lemma</h1>

    <h2>Definición:</h2>
    <p>
        El <span class="highlight">Handshaking Lemma</span> es un principio fundamental en la teoría de grafos que establece que la suma de los grados de todos los vértices de un grafo es igual al doble del número de aristas en el grafo. Matemáticamente, si <em>G = (V, E)</em> es un grafo con <em>V</em> vértices y <em>E</em> aristas, el lema se expresa como:
    </p>

    <div class="math">
        $$ \\sum_{v \\in V} d_G(v) = 2 \\times |E| $$
    </div>

    <p>
        donde <em>d_G(v)</em> representa el grado del vértice <em>v</em>, es decir, el número de aristas conectadas a <em>v</em>.
    </p>

    <h2>Aplicación:</h2>
    <p>El <span class="highlight">Handshaking Lemma</span> tiene varias aplicaciones prácticas en el análisis de grafos:</p>
    <ul>
        <li><strong>Cálculo del Número de Aristas:</strong> Puedes usar el lema para determinar el número de aristas en un grafo simplemente sumando los grados de todos los vértices y dividiendo por 2.</li>
        <li><strong>Grafos Regulares:</strong> En un grafo <em>r</em>-regular, donde cada vértice tiene el mismo grado <strong>r</strong>, el lema permite calcular el número de aristas usando la fórmula
        <script type="math/tex"> |E| = \\frac{nr}{2} </script>, donde <em>n</em> es el número de vértices.</li>
        <li><strong>Paridad de Grados Impares:</strong> Una consecuencia del lema es que en cualquier grafo, el número de vértices con grado impar debe ser siempre par, ya que la suma de los grados debe ser un número par (al ser el doble del número de aristas).</li>
    </ul>

    <h2>Fórmula:</h2>
    <div class="math">
        <script type="math/tex">
        \\sum_{v \\in V} d_G(v) = 2 \\times |E|
        </script>
    </div>

    <h2>Aplicaciones Específicas:</h2>
    <p>
        <strong>Grafo Completo:</strong> En un grafo completo <em>K_n</em>, donde todos los vértices están conectados entre sí, el número de aristas es:
        <div class="math">
        <script type="math/tex">
        \\frac{n \\times (n-1)}{2}
        </script>
        </div>
        derivado directamente del <span class="highlight">Handshaking Lemma</span>.
    </p>

    <p>Esta fórmula y el lema son esenciales para la comprensión y el análisis de estructuras de grafos en diversas áreas, como redes, combinatoria, y algoritmos.</p>
</body>
</html>
"""

# Mostrar el contenido HTML en Google Colab
display(HTML(html_content))
```

```python
import networkx as nx
import matplotlib.pyplot as plt

# Crear un grafo 2-regular en forma de anillo con 6 nodos
n = 6  # Número de nodos
G = nx.cycle_graph(n)  # Grafo en anillo

# Dibujar el grafo
plt.figure(figsize=(6, 6))
pos = nx.circular_layout(G)  # Posicionar los nodos en forma circular
nx.draw(G, pos, with_labels=True, node_color='lightblue', node_size=500, font_size=10, font_weight='bold')
plt.title("Topología de Anillo (2-Regular)")
plt.show()
```

```python
import networkx as nx
import matplotlib.pyplot as plt

# Crear un hipercubo de dimensión 3 (8 nodos)
d = 3  # Dimensión del hipercubo
G = nx.hypercube_graph(d)

# Dibujar el grafo
plt.figure(figsize=(6, 6))
pos = nx.spring_layout(G)  # Posicionar los nodos
nx.draw(G, pos, with_labels=True, node_color='lightgreen', node_size=500, font_size=10, font_weight='bold')
plt.title(f"Topología de Hipercubo (Dimensión {d})")
plt.show()
```

```python
import networkx as nx
import matplotlib.pyplot as plt

# Crear una red de malla de 3x3 nodos
rows, cols = 3, 3  # Dimensiones de la malla
G = nx.grid_2d_graph(rows, cols)  # Crear grafo de malla 2D

# Redibujar los nodos de la malla en un formato más visual
pos = {(x, y): (y, -x) for x, y in G.nodes()}  # Disposición de la malla

# Dibujar el grafo
plt.figure(figsize=(6, 6))
nx.draw(G, pos, with_labels=True, node_color='lightcoral', node_size=500, font_size=10, font_weight='bold')
plt.title(f"Topología de Malla ({rows}x{cols})")
plt.show()
```

```python
import networkx as nx
import matplotlib.pyplot as plt

# Crear una red completamente conectada (clique) con 5 nodos
n = 5  # Número de nodos
G = nx.complete_graph(n)  # Grafo completamente conectado

# Dibujar el grafo
plt.figure(figsize=(6, 6))
pos = nx.spring_layout(G)  # Posicionar los nodos
nx.draw(G, pos, with_labels=True, node_color='lightblue', node_size=500, font_size=10, font_weight='bold')
plt.title("Topología de Malla Completa")
plt.show()
```

```python
import networkx as nx
import matplotlib.pyplot as plt

# Función para crear un grafo r-regular
def grafo_regular(n, r):
    """
    Crea un grafo r-regular con n vértices.

    Args:
    - n: número de vértices
    - r: grado regular (cada vértice tiene grado r)

    Returns:
    - G: grafo r-regular
    """
    # Crear el grafo r-regular
    G = nx.random_regular_graph(r, n)

    # Dibujar el grafo
    nx.draw(G, with_labels=True, node_color="lightblue", node_size=500, font_weight="bold")
    plt.title(f"Grafo {r}-regular con {n} vértices")
    plt.show()

    return G

# Aplicar el Handshaking Lemma y mostrar la suma de grados y el número de aristas
def aplicar_handshaking_lemma(G):
    """
    Aplica el Handshaking Lemma para verificar que la suma de los grados es el doble del número de aristas.

    Args:
    - G: grafo (de NetworkX)
    """
    # Número de aristas
    num_aristas = G.number_of_edges()

    # Obtener los grados de los vértices
    grados = dict(G.degree())

    # Mostrar el grado de cada nodo
    print("Grados de los vértices:")
    for nodo, grado in grados.items():
        print(f"Vértice {nodo}: Grado {grado}")

    # Suma de los grados de los vértices
    suma_grados = sum(grados.values())

    # Mostrar los resultados
    print(f"\nNúmero de aristas (|E|): {num_aristas}")
    print(f"Suma de los grados de los vértices: {suma_grados}")

    # Verificación del Handshaking Lemma
    if suma_grados == 2 * num_aristas:
        print("El Handshaking Lemma se cumple: La suma de los grados es igual al doble del número de aristas.")
    else:
        print("Error: El Handshaking Lemma no se cumple.")

# Parámetros del grafo regular
n = 6  # Número de vértices
r = 3  # Grado regular

# Crear el grafo r-regular
G = grafo_regular(n, r)

# Aplicar el Handshaking Lemma
aplicar_handshaking_lemma(G)

```

```python
import networkx as nx
import matplotlib.pyplot as plt
from IPython.display import display, clear_output
import ipywidgets as widgets

# Definimos los datos del grafo
vertices = ['P', 'Q', 'R', 'S', 'T']
aristas = [('P', 'Q'), ('P', 'R'), ('Q', 'R'), ('Q', 'S'), ('S', 'T')]
numero_aristas = len(aristas)

# Grados correctos de cada vértice
grados_correctos = {
    'P': 2,
    'Q': 3,
    'R': 2,
    'S': 2,
    'T': 1
}

# Función para mostrar el grafo
def mostrar_grafo(vertices, aristas):
    G = nx.Graph()
    G.add_nodes_from(vertices)
    G.add_edges_from(aristas)

    # Dibujar el grafo
    pos = nx.spring_layout(G)
    nx.draw(G, pos, with_labels=True, node_color='lightblue', node_size=2000, font_size=12, font_weight='bold')
    nx.draw_networkx_edges(G, pos, edgelist=aristas)

    plt.title("Grafo de los vértices y aristas")
    plt.show()

# Mostrar las aristas
print("El grafo tiene las siguientes aristas:")
for arista in aristas:
    print(f"{arista[0]} - {arista[1]}")

# Dibujar el grafo visualmente
mostrar_grafo(vertices, aristas)

# Crear un diccionario para los widgets de entrada de los grados
grados_widgets = {vertice: widgets.BoundedIntText(min=0, max=10, description=f'{vertice}:', value=0) for vertice in vertices}

# Crear un botón para validar la entrada con un color llamativo
boton_validar = widgets.Button(description="Validar Grados", button_style='info')  # Cambiado a color azul

# Crear un botón para jugar nuevamente
boton_reiniciar = widgets.Button(description="Jugar nuevamente", button_style='success')

# Contenedor para mostrar resultados
resultados_output = widgets.Output()

# Función para validar los grados
def validar_grados(b):
    with resultados_output:
        clear_output()  # Limpiar la salida previa
        aciertos = 0
        errores = 0
        suma_grados_ingresados = 0
        for vertice in vertices:
            grado_ingresado = grados_widgets[vertice].value
            grado_correcto = grados_correctos[vertice]
            suma_grados_ingresados += grado_ingresado
            if grado_ingresado == grado_correcto:
                print(f"¡Correcto! El grado de {vertice} es {grado_correcto}.")
                aciertos += 1
            else:
                print(f"Incorrecto. El grado de {vertice} debería ser {grado_correcto}, pero ingresaste {grado_ingresado}.")
                errores += 1

        # Mostrar el número total de aciertos
        print(f"\nTotal de aciertos: {aciertos} de {len(vertices)}")

        # Calcular la suma teórica según el lema de Handshaking
        suma_grados_teorica = 2 * numero_aristas

        print(f"\nSuma de los grados ingresados: {suma_grados_ingresados}")
        print(f"Suma teórica según el lema: {suma_grados_teorica}")

        # Mostrar mensaje sobre la suma de grados
        if suma_grados_ingresados == suma_grados_teorica:
            if errores == 0:
                print("¡Correcto! La suma de los grados y los grados individuales coinciden con el Lema de Handshaking.")
            else:
                print("La suma de los grados coincide con el Lema de Handshaking, **pero** algunos grados individuales son incorrectos.")
        else:
            print("Error: la suma de los grados no coincide con el Lema de Handshaking.")

# Función para reiniciar el juego
def reiniciar_juego(b):
    # Reiniciar los valores de los widgets de entrada a 0
    for widget in grados_widgets.values():
        widget.value = 0
    with resultados_output:
        clear_output()  # Limpiar los resultados previos
        print("\n¡Juego reiniciado! Puedes ingresar los grados nuevamente.\n")

# Conectar los botones con las funciones
boton_validar.on_click(validar_grados)
boton_reiniciar.on_click(reiniciar_juego)

# Mostrar los widgets de entrada y los botones
display(*grados_widgets.values())
display(boton_validar)
display(boton_reiniciar)
display(resultados_output)
```

```python
import networkx as nx
import matplotlib.pyplot as plt

# Definición del grafo
graph = {
    'a': ['b', 'c', 'd'],
    'b': ['a', 'c'],
    'c': ['a', 'b', 'd'],
    'd': ['a', 'c']
}

# Función para convertir el diccionario a un objeto de NetworkX
def create_graph(graph_dict):
    G = nx.Graph()
    for node in graph_dict:
        for neighbor in graph_dict[node]:
            G.add_edge(node, neighbor)
    return G

# Crear el grafo
G = create_graph(graph)

# Definir las secuencias de vértices
walk = ['a', 'b', 'c', 'a', 'd']  # Caminata simple (no cerrada)
path = ['a', 'b', 'c', 'd']  # Camino
closed_walk_not_cycle = ['a', 'b', 'c', 'a', 'c', 'd', 'a']  # Caminata cerrada que NO es un ciclo
cycle = ['a', 'b', 'c', 'd', 'a']  # Ciclo

# Función para dibujar el grafo con la secuencia resaltada
def draw_graph(G, path=None, title="Grafo", edge_color='orange', node_start_color='green', node_end_color='red'):
    plt.figure(figsize=(10, 8))
    pos = nx.spring_layout(G, seed=42)  # Semilla para mantener la disposición de nodos constante
    nx.draw(G, pos, with_labels=True, node_color='lightblue', node_size=700, edge_color='gray', width=2, font_size=14, font_weight='bold')

    if path:
        edge_path = [(path[i], path[i + 1]) for i in range(len(path) - 1)]
        nx.draw_networkx_edges(G, pos, edgelist=edge_path, edge_color=edge_color, width=4)
        nx.draw_networkx_nodes(G, pos, nodelist=[path[0]], node_color=node_start_color, node_size=800)
        nx.draw_networkx_nodes(G, pos, nodelist=[path[-1]], node_color=node_end_color, node_size=800)
        edge_labels = {(path[i], path[i + 1]): f"{path[i]}->{path[i + 1]}" for i in range(len(path) - 1)}
        nx.draw_networkx_edge_labels(G, pos, edge_labels=edge_labels, font_color='black', font_size=12)

    plt.title(title, size=20)
    plt.show()

# Visualizaciones
draw_graph(G, walk, title="Caminata: a -> b -> c -> a -> d", edge_color='orange', node_start_color='yellow', node_end_color='yellow')
draw_graph(G, path, title="Camino: a -> b -> c -> d", edge_color='blue', node_start_color='green', node_end_color='red')
draw_graph(G, closed_walk_not_cycle, title="Caminata Cerrada (no es ciclo): a -> b -> c -> a -> c -> d -> a", edge_color='purple', node_start_color='yellow', node_end_color='yellow')
draw_graph(G, cycle, title="Ciclo: a -> b -> c -> d -> a", edge_color='green', node_start_color='orange', node_end_color='orange')
```

```python
import networkx as nx
import matplotlib.pyplot as plt

# Crear el grafo G
G = nx.Graph()
edges = [('a', 'b'), ('b', 'c'), ('c', 'd'), ('d', 'e'), ('a', 'd')]
G.add_edges_from(edges)

# Caminata más corta W1: a -> d -> e
shortest_path_W1 = ['a', 'd', 'e']
shortest_path_length = len(shortest_path_W1) - 1  # Número de pasos en la caminata más corta

# Caminata más larga W2: a -> b -> c -> d -> e
longer_path_W2 = ['a', 'b', 'c', 'd', 'e']
longer_path_length = len(longer_path_W2) - 1  # Número de pasos en la caminata más larga

# Dibujar el grafo
pos = nx.spring_layout(G, seed=42)  # Semilla para disposición constante
plt.figure(figsize=(10, 8))
nx.draw(G, pos, with_labels=True, node_color='lightblue', node_size=500, edge_color='gray', width=2)

# Dibujar la caminata más corta en naranja con línea más gruesa
nx.draw_networkx_edges(G, pos, edgelist=[(shortest_path_W1[i], shortest_path_W1[i+1]) for i in range(len(shortest_path_W1)-1)],
                       edge_color='orange', width=6, alpha=0.7, style='solid')

# Etiquetas para la caminata más corta
for i in range(len(shortest_path_W1) - 1):
    edge_label = f"{i+1}"
    nx.draw_networkx_edge_labels(G, pos, edge_labels={(shortest_path_W1[i], shortest_path_W1[i+1]): edge_label},
                                 font_color='black', font_size=12, verticalalignment='bottom')

# Dibujar la caminata más larga en azul con línea discontinua
nx.draw_networkx_edges(G, pos, edgelist=[(longer_path_W2[i], longer_path_W2[i+1]) for i in range(len(longer_path_W2)-1)],
                       edge_color='blue', width=4, alpha=0.7, style='dashed')

# Etiquetas para la caminata más larga
for i in range(len(longer_path_W2) - 1):
    edge_label = f"{i+1}"
    nx.draw_networkx_edge_labels(G, pos, edge_labels={(longer_path_W2[i], longer_path_W2[i+1]): edge_label},
                                 font_color='red', font_size=12, verticalalignment='top')

# Añadir una leyenda
plt.legend(["Caminata más corta (naranja)", "Caminata más larga (azul)"], loc="upper left")

# Añadir título y mostrar la gráfica
plt.title("Comparación de Caminatas en un Grafo")
plt.show()

# Imprimir el número de pasos en cada caminata
print(f"Número de pasos en la caminata más corta: {shortest_path_length}")
print(f"Número de pasos en la caminata más larga: {longer_path_length}")
```

```python
import networkx as nx
import matplotlib.pyplot as plt

# Crear un grafo desconexo
G_desconexo = nx.Graph()
edges = [('A', 'B'), ('B', 'C'), ('D', 'E')]  # El grupo (A, B, C) no está conectado con (D, E)
G_desconexo.add_edges_from(edges)

# Verificar si el grafo es conexo
print("¿El grafo es conexo?", nx.is_connected(G_desconexo))

# Dibujar el grafo
plt.figure(figsize=(8, 6))
nx.draw(G_desconexo, with_labels=True, node_color='lightcoral', node_size=500, edge_color='gray', width=2)
plt.title("Grafo Desconexo")
plt.show()
```

```python
import networkx as nx
import matplotlib.pyplot as plt

# Crear un grafo desconexo con múltiples componentes
G_multi_comps = nx.Graph()
edges = [('A', 'B'), ('B', 'C'), ('D', 'E'), ('F', 'G'), ('G', 'H')]
G_multi_comps.add_edges_from(edges)

# Encontrar y listar los componentes del grafo
components = list(nx.connected_components(G_multi_comps))
print("Componentes del grafo:", components)

# Dibujar el grafo con diferentes colores para cada componente
plt.figure(figsize=(8, 6))
pos = nx.spring_layout(G_multi_comps)
colors = ['lightblue', 'lightgreen', 'lightcoral']

for i, component in enumerate(components):
    nx.draw_networkx_nodes(G_multi_comps, pos, nodelist=component, node_color=colors[i])
    nx.draw_networkx_edges(G_multi_comps, pos, edgelist=[(u, v) for u in component for v in G_multi_comps.neighbors(u) if v in component], edge_color='gray', width=2)

nx.draw_networkx_labels(G_multi_comps, pos)
plt.title("Componentes del Grafo")
plt.show()
```

```python
import networkx as nx
import matplotlib.pyplot as plt
import matplotlib.colors as mcolors

# Crear un grafo desconexo con varios componentes
G = nx.Graph()
edges = [('A', 'B'), ('B', 'C'), ('D', 'E'), ('E', 'F'), ('G', 'H'), ('H', 'I'), ('I', 'G')]
G.add_edges_from(edges)

# Identificar los componentes conexos
components = list(nx.connected_components(G))

# Crear una figura para la visualización
plt.figure(figsize=(10, 8))
pos = nx.spring_layout(G, seed=42)  # Usar un layout de resorte para mejor visualización

# Asignar colores diferentes a cada componente utilizando una paleta de colores
colors = list(mcolors.TABLEAU_COLORS.values())  # Colores predefinidos en matplotlib
for i, component in enumerate(components):
    # Dibujar nodos para cada componente
    nx.draw_networkx_nodes(G, pos, nodelist=list(component), node_color=colors[i % len(colors)], node_size=700, label=f'Componente {i+1}')
    # Dibujar aristas dentro del componente
    nx.draw_networkx_edges(G, pos, edgelist=[edge for edge in G.edges if edge[0] in component and edge[1] in component], edge_color=colors[i % len(colors)], width=2)

# Dibujar etiquetas de nodos
nx.draw_networkx_labels(G, pos, font_size=14, font_color='black')

# Mejorar la leyenda y el título
plt.title("Componentes de un Grafo Desconexo", fontsize=18, fontweight='bold')
plt.legend(scatterpoints=1, loc="upper right", fontsize=12, title="Componentes", title_fontsize='13')
plt.axis('off')  # Quitar los ejes para una visualización más limpia
plt.show()

# Imprimir información de los componentes de manera clara
print("Información de los Componentes del Grafo:")
for i, component in enumerate(components):
    subgraph = G.subgraph(component)
    print(f"\nComponente {i+1}:")
    print(f"  - Nodos: {sorted(component)}")
    print(f"  - Subgrafo es conexo: {nx.is_connected(subgraph)}")
    print(f"  - Número de nodos: {subgraph.number_of_nodes()}")
    print(f"  - Número de aristas: {subgraph.number_of_edges()}")
```

```python
from IPython.display import display, HTML

html_content = """
<!DOCTYPE html>
<html lang="es">
<head>
    <meta charset="UTF-8">
    <meta name="viewport" content="width=device-width, initial-scale=1.0">
    <title>Grafos Eulerianos</title>
    <style>
        body {
            font-family: Arial, sans-serif;
            line-height: 1.6;
            background-color: #f9f9f9;
            margin: 0;
            padding: 20px;
            color: #333;
        }
        h1, h2, h3 {
            color: #2c3e50;
            margin-bottom: 15px;
        }
        h1 {
            text-align: center;
            font-size: 2.2em;
            margin-top: 0;
        }
        h2 {
            font-size: 1.8em;
            margin-top: 30px;
        }
        h3 {
            font-size: 1.5em;
            margin-top: 25px;
        }
        p, ul {
            background: #ffffff;
            padding: 15px;
            border-radius: 5px;
            box-shadow: 0 0 10px rgba(0, 0, 0, 0.1);
            margin-bottom: 20px;
        }
        ul {
            list-style-type: disc;
            padding-left: 40px;
        }
        li {
            margin-bottom: 10px;
        }
        .example {
            background-color: #e8f0fe;
            padding: 10px;
            border-left: 4px solid #1e88e5;
            border-radius: 5px;
            font-style: italic;
            margin-top: -10px;
            margin-bottom: 20px;
        }
        .summary {
            background-color: #f4e3b2;
            padding: 10px;
            border-left: 4px solid #f1c40f;
            border-radius: 5px;
            font-weight: bold;
        }
    </style>
</head>
<body>

    <h1>Grafos Eulerianos</h1>

    <h2>1. Grado Par y Grafos Eulerianos</h2>
    <p>
        <strong>Definición:</strong> Un multigrafo (un grafo que puede tener más de una arista entre dos vértices) es Euleriano si todos sus vértices tienen grado par. El grado de un vértice es el número de aristas que lo conectan a otros vértices.
    </p>
    <p class="example">
        <strong>Implicación:</strong> Si todos los vértices de un grafo tienen un número par de aristas conectadas a ellos, el grafo tiene la posibilidad de contener un ciclo Euleriano.
    </p>

    <h2>2. Caminos y Ciclos en Grafos Eulerianos</h2>
    <ul>
        <li><strong>Sendero:</strong> Un sendero en un grafo es una caminata que no repite aristas. Es decir, puedes moverte de un vértice a otro sin recorrer la misma arista más de una vez.</li>
        <li><strong>Recorrido:</strong> Un recorrido es una caminata cerrada (empieza y termina en el mismo vértice) que no repite aristas.</li>
        <li><strong>Ciclo Euleriano:</strong> Un ciclo Euleriano (también llamado Eulerian tour) es un recorrido que contiene todas las aristas del grafo exactamente una vez y termina en el vértice donde comenzó.</li>
        <li><strong>Sendero Euleriano:</strong> Un sendero Euleriano es un sendero que contiene todas las aristas del grafo, pero no necesariamente termina en el vértice donde comenzó.</li>
    </ul>

    <h2>3. Condición para la Existencia de un Ciclo Euleriano</h2>
    <p>
        Un grafo tiene un ciclo Euleriano si y solo si:
    </p>
    <ul>
        <li>Todos los vértices del grafo tienen grado par.</li>
        <li>El grafo es conexo (todos los vértices están conectados en un único componente, ignorando las aristas).</li>
    </ul>

    <h2>4. Ejemplo Práctico de un Grafo Euleriano</h2>
    <p>
        Considera un grafo con los vértices <i>v<sub>1</sub></i>, <i>v<sub>2</sub></i>, <i>v<sub>3</sub></i>, <i>v<sub>4</sub></i>, <i>v<sub>5</sub></i>, <i>v<sub>6</sub></i> y las aristas que conectan estos vértices de la siguiente manera:
    </p>
    <div class="example">
        <pre>
v<sub>1</sub> -- v<sub>2</sub> -- v<sub>3</sub>
|     |     |
v<sub>6</sub> -- v<sub>5</sub> -- v<sub>4</sub>
        </pre>
    </div>
    <p>
        <strong>Eulerian Tour en el Ejemplo:</strong> Un ejemplo de recorrido Euleriano en este grafo sería:
    </p>
    <p class="example">
        v<sub>1</sub> &rarr; v<sub>2</sub> &rarr; v<sub>3</sub> &rarr; v<sub>4</sub> &rarr; v<sub>5</sub> &rarr; v<sub>6</sub> &rarr; v<sub>1</sub> &rarr; v<sub>5</sub> &rarr; v<sub>2</sub> &rarr; v<sub>4</sub> &rarr; v<sub>1</sub>
    </p>
    <p>
        Este recorrido empieza y termina en <i>v<sub>1</sub></i>, recorriendo todas las aristas del grafo exactamente una vez.
    </p>

    <h2>5. Dibujo del Ciclo Euleriano</h2>
    <p>
        El concepto de un ciclo Euleriano tiene una aplicación visual interesante: un grafo tiene un ciclo Euleriano si puede ser dibujado en un papel sin levantar el lápiz y sin recorrer la misma arista más de una vez.
    </p>
    <p class="example">
        <strong>Ejemplo de Dibujo:</strong> Imagina que dibujas el grafo dado en un papel comenzando en <i>v<sub>1</sub></i>, y sigues el ciclo Euleriano mencionado. Podrás recorrer todas las conexiones entre los vértices sin levantar el lápiz y sin pasar dos veces por la misma conexión.
    </p>

    <h2>Conclusión</h2>
    <p class="summary">
        Los grafos Eulerianos son un concepto fundamental en la teoría de grafos, con aplicaciones que van desde el diseño de circuitos hasta la planificación de rutas eficientes. La propiedad clave es que todos los vértices tienen grado par, lo que permite recorrer todas las aristas sin repetir ninguna y regresar al punto de inicio.
    </p>

</body>
</html>
"""

# Mostrar el contenido HTML
display(HTML(html_content))
```

```python
# Importar las librerías necesarias
import networkx as nx
import matplotlib.pyplot as plt

# Paso 1: Crear el grafo usando NetworkX
G = nx.Graph()

# Añadir las aristas del grafo según la imagen
edges = [(1, 2), (1, 4), (1, 5), (1, 6),
         (2, 3), (2, 4), (2, 5),
         (3, 4),
         (4, 5),
         (5, 6)]

G.add_edges_from(edges)

# Explicación sobre grafos Eulerianos
explanation = """
Un multigrafo es Euleriano si todos sus vértices tienen un grado par.
Un 'camino' en un grafo es una secuencia de vértices donde no se repite ninguna arista,
mientras que un 'circuito' es un camino cerrado donde no se repite ninguna arista.

Un tour Euleriano en un grafo G es un circuito que contiene todas las aristas de G.
Es decir, pasa por todas las aristas exactamente una vez y regresa al vértice inicial.
Un trail Euleriano es una caminata abierta que cubre todas las aristas sin repetir ninguna.

Un grafo tiene un tour Euleriano si:
1. Todos sus vértices tienen grado par.
2. El grafo es conexo.

Veamos si el grafo cumple estas condiciones.
"""
print(explanation)

# Paso 2: Verificar si el grafo es Euleriano (todos los vértices tienen grado par y es conexo)
def verify_eulerian_properties(G):
    is_connected = nx.is_connected(G)  # Verificar si el grafo es conexo
    all_even_degree = all(degree % 2 == 0 for node, degree in G.degree())  # Verificar si todos los vértices tienen grado par

    return is_connected, all_even_degree

# Verificación de las propiedades Eulerianas
is_connected, all_even_degree = verify_eulerian_properties(G)

if is_connected and all_even_degree:
    print("El grafo es Euleriano. Todos los vértices tienen grado par y el grafo es conexo.")
    print("Por lo tanto, existe un tour Euleriano.")

    # Encontrar el tour Euleriano
    eulerian_tour = list(nx.eulerian_circuit(G))
    print("El tour Euleriano es:")
    print([edge[0] for edge in eulerian_tour] + [eulerian_tour[-1][1]])  # Mostrar el tour Euleriano como lista de nodos

    # Visualizar el grafo y el tour Euleriano
    def draw_graph_with_eulerian_tour(G, eulerian_tour):
        pos = nx.spring_layout(G)  # Disposición de los nodos
        plt.figure(figsize=(8, 6))

        # Dibujar el grafo
        nx.draw(G, pos, with_labels=True, node_color='lightblue', node_size=500, font_size=10, font_weight='bold')

        # Dibujar las aristas del tour Euleriano
        edge_list = [(u, v) for u, v in eulerian_tour]
        nx.draw_networkx_edges(G, pos, edgelist=edge_list, edge_color='r', width=2)

        # Mostrar la gráfica
        plt.title("Grafo con Tour Euleriano resaltado", fontsize=15)
        plt.show()

    draw_graph_with_eulerian_tour(G, eulerian_tour)

else:
    if not is_connected:
        print("El grafo NO es conexo.")
    if not all_even_degree:
        print("No todos los vértices tienen grado par.")
    print("Por lo tanto, NO existe un tour Euleriano.")
```
