```python
from IPython.display import display, HTML

# Contenido HTML Mejorado con colores y mejor visualización
html_content = r"""
<!DOCTYPE html>
<html lang="es">
<head>
    <meta charset="UTF-8">
    <meta name="viewport" content="width=device-width, initial-scale=1.0">
    <title>Relación entre Árbol, Árbol Generador y BFS</title>
    <script type="text/javascript" async
      src="https://cdnjs.cloudflare.com/ajax/libs/mathjax/2.7.7/MathJax.js?config=TeX-MML-AM_CHTML">
    </script>
    <style>
        body {
            font-family: Arial, sans-serif;
            line-height: 1.6;
            background-color: #fefefe;
            margin: 0;
            padding: 20px;
        }
        h1, h2, h3 {
            color: #2c3e50;
        }
        p {
            background: #ffffff;
            padding: 15px;
            border-left: 4px solid #3498db;
            margin-bottom: 15px;
            line-height: 1.8;
        }
        ul, ol {
            background: #ffffff;
            padding: 15px;
            border-left: 4px solid #3498db;
            margin-bottom: 15px;
        }
        li {
            margin-bottom: 10px;
        }
        .highlight {
            background-color: #ffeb3b;
            padding: 2px 5px;
            border-radius: 3px;
            font-weight: bold;
        }
        .important {
            background-color: #f1c40f;
            color: #2c3e50;
            padding: 2px 5px;
            font-weight: bold;
        }
        .example {
            background: #e8f8f5;
            padding: 15px;
            border-left: 4px solid #1abc9c;
            margin-bottom: 15px;
            font-style: italic;
        }
        code {
            background: #ecf0f1;
            padding: 2px 4px;
            font-size: 1em;
        }
        a {
            color: #2980b9;
            text-decoration: none;
        }
        a:hover {
            text-decoration: underline;
        }
    </style>
</head>
<body>
    <h1>Relación entre Árbol, Árbol Generador y BFS</h1>

    <h2>1. Árbol</h2>
    <p>
        Un <span class="important">árbol</span> es un tipo especial de grafo que es conexo y sin ciclos. Un árbol tiene las siguientes propiedades:
    </p>
    <ol>
        <li>Todos los nodos están conectados entre sí a través de un único camino.</li>
        <li>No tiene ciclos.</li>
        <li>Tiene <span class="highlight">n-1</span> aristas, donde <span class="highlight">n</span> es el número de nodos.</li>
    </ol>

    <h2>2. Árbol Generador</h2>
    <p>
        Un <span class="important">árbol generador</span> (o spanning tree en inglés) es un subgrafo de un grafo <strong>G</strong> que:
    </p>
    <ul>
        <li>Contiene todos los nodos de <strong>G</strong>.</li>
        <li>Es conexo y sin ciclos.</li>
        <li>Usa el mínimo número de aristas, lo que es <span class="highlight">n-1</span> aristas, si <strong>G</strong> tiene <strong>n</strong> nodos.</li>
    </ul>
    <p>
        En resumen, un árbol generador es una versión eficiente del grafo que conecta todos los nodos sin formar ciclos.
    </p>

    <h2>3. BFS (Breadth-First Search)</h2>
    <p>
        El <span class="important">BFS</span> es un algoritmo que explora sistemáticamente los nodos de un grafo. Comienza en un nodo inicial y explora todos los nodos vecinos antes de pasar a los vecinos de esos vecinos, siguiendo un enfoque por niveles.
    </p>

    <h2>Relación entre Árbol Generador y BFS</h2>
    <p class="example">
        Cuando ejecutas BFS en un grafo desde un nodo inicial, el algoritmo genera un árbol de búsqueda, que es un <span class="important">árbol generador</span> del grafo. Este árbol conecta todos los nodos accesibles, sin ciclos, y explora los nodos en capas, asegurando que los más cercanos al nodo inicial se procesen primero.
    </p>

    <h2>Ejemplo de uso</h2>
    <p class="example">
        Imagina un grafo que representa una red de caminos entre ciudades. Al aplicar BFS desde una ciudad inicial, obtenemos un <span class="highlight">árbol generador</span> que representa las rutas más cortas para llegar a cada ciudad sin formar ciclos.
    </p>
</body>
</html>
"""

# Mostrar el contenido HTML
display(HTML(html_content))
```

```python
import networkx as nx
import matplotlib.pyplot as plt
from matplotlib import animation
from mpl_toolkits.mplot3d import Axes3D
from IPython.display import HTML
from google.colab import files  # Para descargar archivos en Colab
import zipfile  # Para empaquetar los archivos en ZIP

# Crear el grafo cubo usando números
G = nx.Graph()
V = range(8)  # Nodos numerados de 0 a 7
G.add_nodes_from(V)
E = [
    (0, 1), (0, 2), (0, 4),
    (1, 3), (1, 5),
    (2, 3), (2, 6),
    (4, 5), (4, 6),
    (3, 7), (5, 7), (6, 7)
]
G.add_edges_from(E)

# Algoritmo BFS para recorrer el grafo
def bfs(graph, start):
    visited = set()
    queue = [start]
    bfs_edges = []
    bfs_nodes = [start]

    while queue:
        node = queue.pop(0)
        if node not in visited:
            visited.add(node)
            neighbors = sorted(list(graph.neighbors(node)))  # Ordenar los vecinos en orden creciente
            for neighbor in neighbors:
                if neighbor not in visited:
                    queue.append(neighbor)
                    bfs_edges.append((node, neighbor))  # Registrar las aristas recorridas
                    bfs_nodes.append(neighbor)
    return bfs_nodes, bfs_edges

# Ejecutar BFS comenzando desde el nodo 0
bfs_nodes, bfs_edges = bfs(G, 0)

# Generar posiciones 3D para los nodos
pos_3d = {
    0: (0, 0, 0),
    1: (0, 0, 1),
    2: (0, 1, 0),
    3: (0, 1, 1),
    4: (1, 0, 0),
    5: (1, 0, 1),
    6: (1, 1, 0),
    7: (1, 1, 1)
}

# Crear la figura y el eje 3D
fig = plt.figure(figsize=(8, 8))
ax = fig.add_subplot(111, projection='3d')

# Función para dibujar el grafo cubo en 3D, mostrando BFS dinámicamente sin etiquetas
def draw_graph(ax, G, pos, bfs_edges, bfs_nodes, step):
    ax.clear()

    # Dibujar las aristas solo hasta el paso actual
    for edge in bfs_edges[:step]:
        x = [pos[edge[0]][0], pos[edge[1]][0]]
        y = [pos[edge[0]][1], pos[edge[1]][1]]
        z = [pos[edge[0]][2], pos[edge[1]][2]]
        ax.plot(x, y, z, color='red', linewidth=3)  # Aristas recorridas en rojo

    # Dibujar nodos visitados hasta el paso actual
    for node in bfs_nodes[:step]:
        ax.scatter(pos[node][0], pos[node][1], pos[node][2], color='yellow', s=900, edgecolors='black', linewidths=2)  # Nodos visitados en amarillo

    # Configurar los límites de los ejes
    ax.set_xlim(-0.5, 1.5)
    ax.set_ylim(-0.5, 1.5)
    ax.set_zlim(-0.5, 1.5)

    # Remover las líneas de los ejes para una estética más limpia
    ax.set_axis_off()

# Dibujar el grafo inicialmente
draw_graph(ax, G, pos_3d, bfs_edges, bfs_nodes, step=0)

# Función de actualización para la animación de BFS
def update_bfs(step):
    draw_graph(ax, G, pos_3d, bfs_edges, bfs_nodes, step)
    ax.view_init(elev=30, azim=step * 10)  # Rotar la vista para mayor dinamismo
    return fig,

# Crear la animación
ani = animation.FuncAnimation(fig, update_bfs, frames=len(bfs_nodes), interval=1000, blit=False)

# Guardar la animación como GIF y MP4
gif_path = 'bfs_cubo_animacion_sin_etiquetas.gif'
mp4_path = 'bfs_cubo_animacion_sin_etiquetas.mp4'

ani.save(gif_path, writer='pillow', fps=1)
ani.save(mp4_path, writer='ffmpeg', fps=1)

print(f"Animaciones guardadas como '{gif_path}' y '{mp4_path}'.")

# Crear un archivo ZIP que contenga tanto el GIF como el MP4
zip_filename = 'bfs_cubo_animacion_sin_etiquetas.zip'
with zipfile.ZipFile(zip_filename, 'w') as zipf:
    zipf.write(gif_path)
    zipf.write(mp4_path)

print(f"Archivos empaquetados en '{zip_filename}'.")

# Descargar el archivo ZIP en Google Colab
files.download(zip_filename)

# Mostrar la animación en el notebook
HTML(ani.to_jshtml())
```

```python
import networkx as nx
import matplotlib.pyplot as plt
from matplotlib import animation
from IPython.display import HTML
from google.colab import files  # Para descargar archivos en Colab
import zipfile

# Crear el grafo
G = nx.Graph()
edges = [(1, 3), (1, 9), (3, 2), (3, 6), (9, 6), (9, 8), (6, 4), (6, 7), (8, 5)]
G.add_edges_from(edges)

# Aplicar BFS desde el nodo 1
bfs_tree = nx.bfs_tree(G, source=1)

# Obtener niveles de BFS (distancias desde el nodo raíz)
levels = nx.single_source_shortest_path_length(G, source=1)

# Definir colores para cada nivel
level_colors = {
    0: 'lightgreen',  # Nivel 0 (nodo raíz)
    1: 'lightblue',   # Nivel 1
    2: 'orange',      # Nivel 2
    3: 'red'          # Nivel 3
}

# Crear una lista de colores para los nodos basados en el nivel
node_colors = [level_colors[levels[node]] for node in G.nodes()]

# Definir las posiciones de los nodos para visualización
pos = {
    1: (0, 3),
    3: (-1, 2),
    9: (1, 2),
    2: (-2, 1),
    6: (0, 1),
    8: (2, 1),
    4: (-1, 0),
    7: (1, 0),
    5: (2, 0)
}

# Crear la figura
fig, ax = plt.subplots(figsize=(8, 6))

# Función para dibujar el grafo completo y resaltar los nodos a medida que BFS avanza
def draw_frame(step):
    ax.clear()  # Limpiar el gráfico en cada frame

    # Dibujar el grafo completo en azul claro (sin resaltar el BFS)
    nx.draw(G, pos, with_labels=True, node_color='lightgrey', edge_color='lightgrey',
            node_size=800, font_size=14, font_weight='bold', ax=ax)

    # Resaltar los nodos y aristas hasta el paso actual del BFS
    nodes_to_draw = list(bfs_tree.nodes())[:step]
    edges_to_draw = list(bfs_tree.edges())[:step-1]

    # Colorear los nodos hasta el paso actual
    node_colors_step = [level_colors[levels[node]] for node in nodes_to_draw]

    # Dibujar los nodos y aristas del árbol BFS hasta el paso actual
    nx.draw_networkx_nodes(bfs_tree, pos, nodelist=nodes_to_draw, node_color=node_colors_step,
                           node_size=800, ax=ax)
    nx.draw_networkx_edges(bfs_tree, pos, edgelist=edges_to_draw, edge_color='black', ax=ax)

    # Etiquetas para los nodos
    nx.draw_networkx_labels(G, pos, ax=ax, font_size=14, font_weight='bold')

    ax.set_title(f"Árbol de Búsqueda en Amplitud (BFS) - Paso {step}")

# Crear la animación
ani = animation.FuncAnimation(fig, draw_frame, frames=len(G.nodes()) + 1, interval=1000, repeat=False)

# Guardar la animación como GIF y MP4
gif_path = 'bfs_animacion.gif'
mp4_path = 'bfs_animacion.mp4'

ani.save(gif_path, writer='pillow', fps=1)
ani.save(mp4_path, writer='ffmpeg', fps=1)

# Crear un archivo ZIP que contenga tanto el GIF como el MP4
zip_filename = 'bfs_animacion.zip'
with zipfile.ZipFile(zip_filename, 'w') as zipf:
    zipf.write(gif_path)
    zipf.write(mp4_path)

print(f"Animaciones guardadas como '{gif_path}' y '{mp4_path}', empaquetadas en '{zip_filename}'.")

# Descargar el archivo ZIP en Google Colab
files.download(zip_filename)

# Mostrar la animación en el notebook
HTML(ani.to_jshtml())
```
