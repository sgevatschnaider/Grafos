```python
from IPython.display import display, HTML

html_content = """
<!DOCTYPE html>
<html lang="es">
<head>
    <meta charset="UTF-8">
    <title>El Grafo de Petersen y el Límite de Ramanujan en Teoría de Grafos</title>
    <style>
        body {
            font-family: 'Arial', sans-serif;
            line-height: 1.6;
            padding: 20px;
            color: #333;
            background-color: #f5f5f5;
        }
        h1, h2, h3, h4 {
            color: #0066cc;
        }
        p {
            margin-bottom: 10px;
        }
        .collapsible {
            background-color: #0066cc;
            color: white;
            cursor: pointer;
            padding: 10px;
            width: 100%;
            text-align: left;
            outline: none;
            font-size: 1em;
            border: none;
            border-radius: 5px;
            margin-top: 10px;
            transition: background-color 0.3s;
        }
        .collapsible:hover {
            background-color: #0056b3;
        }
        .content {
            padding: 0 18px;
            display: none;
            overflow: hidden;
            background-color: #e0e7ef;
            border-radius: 5px;
            margin-bottom: 10px;
        }
        .image-container {
            text-align: center;
            margin: 20px 0;
        }
        .image-container img {
            max-width: 100%;
            height: auto;
            border: 1px solid #ddd;
            border-radius: 5px;
            padding: 5px;
        }
    </style>
</head>
<body>
    <h1>El Grafo de Petersen y el Límite de Ramanujan en Teoría de Grafos</h1>

    <button class="collapsible">4.1 Propiedades del Grafo de Petersen</button>
    <div class="content">
        <p>El <strong>Grafo de Petersen</strong> es un grafo no dirigido con 10 vértices y 15 aristas, conocido por sus interesantes propiedades en teoría de grafos. Es un grafo cúbico (cada vértice tiene grado 3) y altamente simétrico, lo que lo convierte en un objeto de estudio frecuente en la teoría de grafos y en aplicaciones como la criptografía y el diseño de redes robustas.</p>
        <p>Algunas propiedades destacadas del Grafo de Petersen son:</p>
        <ul>
            <li><strong>Conectividad:</strong> Es 3-conexo, lo que significa que se necesitan al menos 3 aristas para desconectarlo.</li>
            <li><strong>No es planar:</strong> No puede ser dibujado en un plano sin que se crucen algunas de sus aristas.</li>
            <li><strong>Es un grafo de expander:</strong> Su estructura hace que cualquier subconjunto de vértices esté fuertemente interconectado con el resto del grafo.</li>
            <li><strong>Diámetro y Girth:</strong> Su diámetro es 2, y su girth (la longitud del ciclo más corto) es 5.</li>
        </ul>
    </div>

    <button class="collapsible">4.2 Espectro de Valores Propios y el Límite de Ramanujan</button>
    <div class="content">
        <p>El <strong>espectro de valores propios</strong> de un grafo se refiere a los valores propios de su matriz de adyacencia, y se utiliza para estudiar las propiedades estructurales y de expansión del grafo. En el caso del Grafo de Petersen, el espectro de sus valores propios es {3, 1, 1, 1, 1, -2, -2, -2, -2, -2}.</p>
        <p><strong>Límite de Ramanujan:</strong> Los grafos que cumplen el Límite de Ramanujan tienen excelentes propiedades de expansión. En este caso, para un grafo cúbico, el Límite de Ramanujan establece que los valores propios no triviales deben estar en el rango [-2√2, 2√2], que es aproximadamente [-2.83, 2.83].</p>
        <p><strong>Interpretación:</strong> El Grafo de Petersen cumple este límite, ya que todos sus valores propios no triviales están dentro de este rango. Esto hace que el Grafo de Petersen sea un buen grafo de expander, es decir, tiene propiedades de expansión óptimas para muchas aplicaciones en redes y criptografía.</p>
    </div>

    <button class="collapsible">4.3 Aplicación del Límite de Ramanujan en Seguridad de Redes y Teoría de Expansores</button>
    <div class="content">
        <p>Los <strong>grafos de expander</strong>, y en particular los grafos que cumplen el Límite de Ramanujan, son de gran interés en aplicaciones de seguridad de redes, ya que tienen propiedades de conectividad que permiten construir redes resilientes y eficientes.</p>
        <ul>
            <li><strong>Redes P2P y Redes de Difusión:</strong> Los grafos de expander aseguran que cualquier nodo de la red esté conectado a varios otros nodos, facilitando la comunicación y redundancia.</li>
            <li><strong>Distribución de Claves en Criptografía:</strong> Al distribuir claves o datos críticos en una red construida sobre un grafo de expander, se mejora la resiliencia frente a fallos y ataques, ya que la red es difícil de desconectar.</li>
            <li><strong>Algoritmos de Criptografía Post-Cuántica:</strong> Algunos algoritmos de criptografía post-cuántica, como los basados en redes lattices, pueden aprovechar grafos de expander para mejorar la seguridad y eficiencia de las operaciones.</li>
        </ul>
    </div>

    <button class="collapsible">4.4 Visualización de Espectros en Grafos: Ejemplo en Python</button>
    <div class="content">
        <p>Para visualizar el espectro de valores propios de un grafo, se puede usar Python y librerías como <strong>NetworkX</strong> y <strong>Matplotlib</strong>. A continuación, se muestra un ejemplo de código en Python para representar el Grafo de Petersen y su espectro de valores propios, con el Límite de Ramanujan destacado.</p>
        <pre>
import networkx as nx
import matplotlib.pyplot as plt
import numpy as np

# Crear el Grafo de Petersen
G = nx.petersen_graph()

# Calcular el espectro de valores propios
eigenvalues = np.linalg.eigvalsh(nx.adjacency_matrix(G).todense())

# Graficar el Grafo de Petersen
plt.figure(figsize=(10, 5))
plt.subplot(121)
nx.draw(G, with_labels=True, node_color='lightblue', edge_color='gray', node_size=500, font_weight='bold')
plt.title("Grafo de Petersen")

# Graficar el espectro de valores propios
plt.subplot(122)
plt.bar(range(len(eigenvalues)), eigenvalues, color='gray')
plt.axhline(y=2.83, color='blue', linestyle='--', label="Límite Ramanujan (+2.83)")
plt.axhline(y=-2.83, color='blue', linestyle='--', label="Límite Ramanujan (-2.83)")
plt.xlabel("Índice de valor propio")
plt.ylabel("Valor propio")
plt.title("Espectro de valores propios")
plt.legend()
plt.show()
        </pre>
        <p>Este ejemplo de código genera una visualización del Grafo de Petersen y su espectro de valores propios, resaltando el Límite de Ramanujan en el espectro.</p>
    </div>

    <button class="collapsible">4.5 Explicación de la Imagen Proporcionada</button>
    <div class="content">
        <p>La imagen muestra dos gráficos que analizan el <strong>Grafo de Petersen</strong> y su <strong>espectro de valores propios</strong> con respecto al <strong>Límite de Ramanujan</strong>:</p>
        <ul>
            <li><strong>Gráfico Izquierdo: Grafo de Petersen</strong><br>
                La estructura de la izquierda representa el Grafo de Petersen. Este es un grafo con 10 vértices (nodos) y 15 aristas (conexiones). Cada vértice está conectado a tres otros vértices, lo que significa que es un grafo cúbico (de grado 3). Los nodos están coloreados en azul claro, y las aristas están representadas en gris. La disposición en la imagen refleja la simetría característica del Grafo de Petersen.
            </li>
            <li><strong>Gráfico Derecho: Espectro de Valores Propios del Grafo de Petersen</strong><br>
                Este gráfico de barras muestra los valores propios del Grafo de Petersen, calculados a partir de su matriz de adyacencia. Los valores propios están en el eje vertical y su índice en el eje horizontal. Los valores mayores y menores están dentro del rango del <strong>Límite de Ramanujan</strong> (±2.83), mostrado con líneas azules punteadas en ±2.83. Esto indica que el Grafo de Petersen cumple el límite, demostrando sus propiedades de expansión óptimas, útiles en aplicaciones de redes y criptografía.
            </li>
        </ul>
    </div>

    <script>
        var coll = document.getElementsByClassName("collapsible");
        for (var i = 0; i < coll.length; i++) {
            coll[i].addEventListener("click", function() {
                this.classList.toggle("active");
                var content = this.nextElementSibling;
                if (content.style.display === "block") {
                    content.style.display = "none";
                } else {
                    content.style.display = "block";
                }
            });
        }
    </script>
</body>
</html>
"""

# Display HTML content in Google Colab
display(HTML(html_content))
```

```python
import networkx as nx
import matplotlib.pyplot as plt
import numpy as np
from matplotlib import animation
from mpl_toolkits.mplot3d import Axes3D
from IPython.display import HTML
from google.colab import files

# Crear el Grafo de Petersen y verificar su estructura
G = nx.petersen_graph()
assert len(G.nodes) == 10, "El grafo debe tener 10 nodos para ser el grafo de Petersen."
assert len(G.edges) == 15, "El grafo debe tener 15 aristas para ser el grafo de Petersen."

# Generar posiciones en 3D para los nodos automáticamente
pos_3d = nx.spring_layout(G, dim=3, seed=42)  # Disposición en 3D automática

# Calcular la matriz de adyacencia y sus valores propios
adj_matrix = nx.adjacency_matrix(G).todense()
eigenvalues = np.sort(np.real(np.linalg.eigvals(adj_matrix)))

# Parámetros de Ramanujan
k = 3  # Grado del grafo (3-regular)
ramanujan_bound = 2 * np.sqrt(k - 1)

# Crear la figura y los ejes
fig = plt.figure(figsize=(14, 7))
ax_graph = fig.add_subplot(121, projection='3d')  # Eje 3D para el grafo
ax_spectrum = fig.add_subplot(122)                # Eje 2D para el espectro de valores propios

# Función para dibujar el Grafo de Petersen en 3D
def draw_graph(ax, pos, angle):
    ax.clear()
    for edge in G.edges():
        x, y, z = zip(pos[edge[0]], pos[edge[1]])
        ax.plot(x, y, z, color='gray', linewidth=1.5)

    # Dibujar los nodos
    xs, ys, zs = zip(*[pos[node] for node in G.nodes()])
    ax.scatter(xs, ys, zs, color='skyblue', s=700, edgecolors='black', linewidths=2)

    # Configuración de los ejes
    ax.set_xlim(-1.5, 1.5)
    ax.set_ylim(-1.5, 1.5)
    ax.set_zlim(-1.5, 1.5)
    ax.set_axis_off()
    ax.view_init(elev=30, azim=angle)

# Función para actualizar el espectro de valores propios
def update_spectrum(ax, frame, eigenvalues, ramanujan_bound):
    ax.clear()
    ax.bar(range(len(eigenvalues)), eigenvalues, color='lightgray')

    # Resaltar el valor propio actual
    if frame < len(eigenvalues):
        color = 'green' if abs(eigenvalues[frame]) <= ramanujan_bound else 'red'
        ax.bar(frame, eigenvalues[frame], color=color)

    # Dibujar las líneas de límite de Ramanujan
    ax.axhline(y=ramanujan_bound, color='blue', linestyle='--', label=f'Límite Ramanujan (+{ramanujan_bound:.2f})')
    ax.axhline(y=-ramanujan_bound, color='blue', linestyle='--', label=f'Límite Ramanujan (-{ramanujan_bound:.2f})')
    ax.set_xlabel("Índice de valor propio")
    ax.set_ylabel("Valor propio")
    ax.set_title("Espectro de valores propios")
    ax.legend()

# Función para actualizar el grafo y el espectro en cada frame de la animación
def update(frame, ax_graph, ax_spectrum, pos, eigenvalues, ramanujan_bound):
    angle = frame * 3  # Rotación progresiva
    draw_graph(ax_graph, pos, angle)
    update_spectrum(ax_spectrum, frame, eigenvalues, ramanujan_bound)

# Crear la animación
frames = len(eigenvalues) + 30  # Mostrar algunos frames adicionales al final
ani = animation.FuncAnimation(fig, update, frames=frames, fargs=(ax_graph, ax_spectrum, pos_3d, eigenvalues, ramanujan_bound), interval=500, blit=False)

# Guardar la animación como GIF y MP4
ani.save('petersen_ramanujan_animation.gif', writer='pillow', fps=10)
ani.save('petersen_ramanujan_animation.mp4', writer='ffmpeg', fps=10)

# Descargar los archivos generados (si estás en Google Colab)
files.download('petersen_ramanujan_animation.gif')
files.download('petersen_ramanujan_animation.mp4')

# Mostrar la animación en el notebook
HTML(ani.to_jshtml())
```

```python
# Importar las bibliotecas necesarias
import matplotlib.pyplot as plt
import networkx as nx
import numpy as np

# Crear el grafo de Petersen
G = nx.petersen_graph()

# Calcular la matriz de adyacencia y sus valores propios
adj_matrix = nx.adjacency_matrix(G).todense()
eigenvalues = np.linalg.eigvals(adj_matrix)
eigenvalues = np.sort(np.real(eigenvalues))

# Verificar la condición de Ramanujan
k = 3  # El grafo de Petersen es 3-regular
ramanujan_bound = 2 * np.sqrt(k - 1)
is_ramanujan = all(abs(ev) <= ramanujan_bound for ev in eigenvalues if ev != k)

# Mostrar resultados en consola
print(f"Espectro del grafo de Petersen: {eigenvalues}")
print(f"Límite de Ramanujan para k={k}: ±{ramanujan_bound:.2f}")
print("¿Es el grafo de Ramanujan?", "Sí" if is_ramanujan else "No")

# Visualización del Grafo de Petersen y Espectro
plt.figure(figsize=(12, 6))

# Subplot 1: Grafo de Petersen con colores en los nodos
plt.subplot(1, 2, 1)
# Disposición de posiciones personalizada similar a la imagen proporcionada
pos = {
    0: (0, 1.5),    # Nodo 1
    1: (-1.4, 0.5), # Nodo 2
    2: (-0.8, -1.2),# Nodo 3
    3: (0.8, -1.2), # Nodo 4
    4: (1.4, 0.5),  # Nodo 5
    5: (0, 0.6),    # Nodo 6
    6: (-0.5, 0.2), # Nodo 7
    7: (-0.4, -0.7),# Nodo 8
    8: (0.4, -0.7), # Nodo 9
    9: (0.5, 0.2)   # Nodo 10
}
plt.axis('off')  # Ocultar los ejes

# Asignar colores a los nodos exteriores e interiores
outer_nodes = [0, 1, 2, 3, 4]
inner_nodes = [5, 6, 7, 8, 9]
nx.draw_networkx_nodes(G, pos, nodelist=outer_nodes, node_color='skyblue', node_size=600, edgecolors='black')
nx.draw_networkx_nodes(G, pos, nodelist=inner_nodes, node_color='salmon', node_size=600, edgecolors='black')
nx.draw_networkx_edges(G, pos, edge_color='black', width=2)
plt.title("Grafo de Petersen", fontsize=14)

# Subplot 2: Espectro de valores propios con el límite de Ramanujan
plt.subplot(1, 2, 2)
plt.bar(range(len(eigenvalues)), eigenvalues, color='lightcoral')
plt.axhline(y=ramanujan_bound, color='blue', linestyle='--', label=f'Límite Ramanujan (+{ramanujan_bound:.2f})')
plt.axhline(y=-ramanujan_bound, color='blue', linestyle='--', label=f'Límite Ramanujan (-{ramanujan_bound:.2f})')
plt.xlabel("Índice de valor propio")
plt.ylabel("Valor propio")
plt.title("Espectro de valores propios")
plt.legend()

# Mostrar la visualización completa
plt.suptitle("Grafo de Petersen y condición de Ramanujan", fontsize=16)
plt.tight_layout(rect=[0, 0, 1, 0.95])
plt.show()
```
