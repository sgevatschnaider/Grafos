```python
# -*- coding: utf-8 -*-
"""
Huginn Latent Reasoning Analysis
Compatible con Google Colab o entorno local con ffmpeg instalado.
"""

# ── 0) Instalar dependencias ──────────────────────────────────────────────────
# Ejecutar en Colab:
!pip install -q transformers accelerate matplotlib scikit-learn ffmpeg-python

# ── 1) Importar librerías ─────────────────────────────────────────────────────
import torch
import numpy as np
import matplotlib.pyplot as plt
import matplotlib as mpl
from transformers import (
    AutoModelForCausalLM, AutoTokenizer, GenerationConfig, TextStreamer
)
from matplotlib.animation import FuncAnimation, FFMpegWriter, PillowWriter
from mpl_toolkits.mplot3d import Axes3D  # noqa: F401
from sklearn.decomposition import PCA
import math

# ── 2) Configurar dispositivo ─────────────────────────────────────────────────
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")

# ── 3) Cargar modelo y tokenizer ──────────────────────────────────────────────
model = AutoModelForCausalLM.from_pretrained(
    "tomg-group-umd/huginn-0125",
    trust_remote_code=True,
    torch_dtype=torch.bfloat16,
    low_cpu_mem_usage=True,
    device_map="auto"
)
tokenizer = AutoTokenizer.from_pretrained("tomg-group-umd/huginn-0125")
model.eval()

# ── 4) Preparar prompt en formato chat ────────────────────────────────────────
from dataclasses import dataclass

@dataclass
class Message:
    role: str
    content: str

system_msg = "You are Huginn, a thoughtful AI assistant developed for deep latent reasoning."
user_msg   = "Claire makes a 3 egg omelet every morning for breakfast. How many dozens of eggs will she eat in 4 weeks?"

messages = [
    Message(role="system", content=system_msg),
    Message(role="user",   content=user_msg)
]

formatted = [
    {"role": "Huginn" if m.role=="assistant" else m.role, "content": m.content}
    for m in messages
]

chat_input = tokenizer.apply_chat_template(
    formatted, tokenize=False, add_generation_prompt=True
)
input_ids = tokenizer.encode(
    chat_input, return_tensors="pt", add_special_tokens=False
).to(device)

# ── 5) Configuración de generación ────────────────────────────────────────────
config = GenerationConfig(
    max_length=1024,
    stop_strings=["<|end_text|>","<|end_turn|>"],
    do_sample=False,
    return_dict_in_generate=True,
    eos_token_id=65505,
    bos_token_id=65504,
    pad_token_id=65509
)
streamer = TextStreamer(tokenizer)

# ── 6) Generar respuesta y decodificar ───────────────────────────────────────
outputs = model.generate(
    input_ids, config, num_steps=64,
    tokenizer=tokenizer, streamer=streamer
)
answer = tokenizer.decode(outputs.sequences[0], skip_special_tokens=True)
print("\nRespuesta de Huginn:\n", answer, "\n")

# ── 7) Calcular estados latentes ──────────────────────────────────────────────
@torch.no_grad()
def compute_latents(model, outputs, num_steps=64):
    emb, _, _ = model.embed_inputs(outputs.sequences)
    s = model.initialize_state(emb, deterministic=False)
    lat = [ model.transformer.ln_f(s).cpu().float().numpy() ]
    for _ in range(num_steps):
        s, _, _ = model.iterate_one_step(emb, s)
        lat.append(model.transformer.ln_f(s).cpu().float().numpy())
    return np.stack(lat)

latents = compute_latents(model, outputs, num_steps=64)  # [steps+1, batch, seq_len, hidden]

# ── 8) Visualizar convergencia global ─────────────────────────────────────────
final = latents[-1]
conv = np.linalg.norm(latents - final, axis=-1).mean(axis=(1,2))

plt.figure(figsize=(8,3))
plt.plot(conv, marker='o', linewidth=1.5)
plt.title('Convergencia promedio del estado latente', fontsize=14)
plt.xlabel('Iteración', fontsize=12)
plt.ylabel('||xᵗ - x*|| promedio', fontsize=12)
plt.grid(alpha=0.3)
plt.tight_layout()
plt.show()

# ── 9) Preparar PCA 2D y 6D ──────────────────────────────────────────────────
steps, _, seq_len, hidden_dim = latents.shape
flat = latents.reshape(-1, hidden_dim)
pca2 = PCA(n_components=2).fit(flat)
lat_pca2 = pca2.transform(flat).reshape(steps, seq_len, 2)

pca6 = PCA(n_components=6).fit(flat)
lat_pca6 = pca6.transform(flat).reshape(steps, seq_len, 6)

# ── 10) Animación Waterfall 3D mejorada ───────────────────────────────────────
fig_wf = plt.figure(figsize=(7,7))
ax_wf  = fig_wf.add_subplot(111, projection='3d')
cmap = plt.cm.viridis

# Mostrar cada 5 tokens para no saturar
token_positions = list(range(0, seq_len, 5))

def init_wf():
    ax_wf.clear()
    ax_wf.set_xlim(-50,50); ax_wf.set_ylim(-50,50); ax_wf.set_zlim(0, seq_len)
    ax_wf.set_xlabel('PCA1', fontsize=12)
    ax_wf.set_ylabel('PCA2', fontsize=12)
    ax_wf.set_zlabel('Posición token', fontsize=12)
    return []

def update_wf(frame):
    ax_wf.view_init(elev=30, azim=45 + frame*3)  # giro de ±180° en 60 frames
    init_wf()
    # Dibujar un planeado semitransparente del paso actual
    ax_wf.text(-45, -45, seq_len*0.9, f'Iteración {frame}', color='black', fontsize=12)
    colors = cmap(frame/steps)
    for pos in token_positions:
        traj = lat_pca2[:, pos]
        z = pos * np.ones(steps)
        ax_wf.plot(traj[:,0], traj[:,1], z,
                   color='gray', alpha=0.2, linewidth=0.5)
        ax_wf.scatter(traj[frame,0], traj[frame,1], z[frame],
                      color=colors, s=20, edgecolor='k')
    return []

ani_wf = FuncAnimation(
    fig_wf, update_wf, init_func=init_wf,
    frames=steps, interval=100, blit=False
)
ani_wf.save('waterfall_dinámico.mp4', writer=FFMpegWriter(fps=10), dpi=200)
ani_wf.save('waterfall_dinámico.gif', writer=PillowWriter(fps=10), dpi=100)

# ── 11) Animación Trayectorias de Tokens Seleccionados ───────────────────────
token_ids   = outputs.sequences[0]
token_texts = [tokenizer.decode(int(t)) for t in token_ids]

# Elegir tokens interesantes (ej. los que más tardan en converger)
interesting = [10, 20, 30, 40, 50]  # ajusta según tu secuencia

fig_tok, axes_tok = plt.subplots(
    len(interesting), 3,
    figsize=(12, 4*len(interesting))
)
plt.tight_layout()
norm = mpl.colors.Normalize(vmin=0, vmax=steps)
sm   = mpl.cm.ScalarMappable(cmap=cmap, norm=norm)
sm.set_array([])

def init_tok():
    for i, idx in enumerate(interesting):
        for j in range(3):
            ax = axes_tok[i,j]
            ax.clear()
            ax.set_xticks([]); ax.set_yticks([])
            if j==0:
                ax.set_title(f'Token "{token_texts[idx]}"', fontsize=12)
    return []

def update_tok(frame):
    init_tok()
    plt.subplots_adjust(hspace=0.4)
    for i, idx in enumerate(interesting):
        for j in range(3):
            ax = axes_tok[i,j]
            c1, c2 = 2*j, 2*j+1
            traj = lat_pca6[:frame+1, idx, [c1,c2]]
            com  = traj.mean(axis=0)
            t    = traj - com
            # Dibujar trayectoria parcial
            ax.plot(t[:,0], t[:,1], color='gray', alpha=0.3, linewidth=1)
            sc = ax.scatter(
                t[:,0], t[:,1],
                c=np.arange(frame+1),
                cmap=cmap, norm=norm,
                s=30, edgecolor='k'
            )
            ax.scatter(0,0, marker='x', c='red', s=80)
    # Añadir colorbar sólo una vez
    if frame==0:
        cbar = fig_tok.colorbar(sm, ax=axes_tok, location='right', fraction=0.02)
        cbar.set_label('Iteración', fontsize=12)
    return []

ani_tok = FuncAnimation(
    fig_tok, update_tok, init_func=init_tok,
    frames=steps, interval=100, blit=False
)
ani_tok.save('trayectorias_dinámico.mp4', writer=FFMpegWriter(fps=10), dpi=200)
ani_tok.save('trayectorias_dinámico.gif', writer=PillowWriter(fps=10), dpi=100)

print(" Animaciones generadas y guardadas:")
print("   • waterfall_mejorado.mp4 / .gif")
print("   • trayectorias_mejorado.mp4 / .gif")
```
