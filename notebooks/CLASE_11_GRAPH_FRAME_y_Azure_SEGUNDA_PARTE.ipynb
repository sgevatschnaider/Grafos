1. Tripleta Abierta
Una tripleta abierta es un conjunto de tres nodos donde dos de ellos están conectados, mientras que el tercero está desconectado de uno de ellos.




```python
import networkx as nx
import matplotlib.pyplot as plt

# Crear un grafo dirigido
G_open = nx.DiGraph()

# Añadir nodos y aristas para representar la tripleta abierta
G_open.add_nodes_from(["A", "B", "C"])
G_open.add_edges_from([("A", "B"), ("B", "C")])

# Dibujar el grafo
pos = nx.spring_layout(G_open)
nx.draw(G_open, pos, with_labels=True, node_color="skyblue", font_weight="bold", node_size=700)
nx.draw_networkx_edge_labels(G_open, pos, edge_labels={("A", "B"): "A→B", ("B", "C"): "B→C"})
plt.title("Tripleta Abierta (A→B y B→C, pero A y C no están conectados)")
plt.show()
```

2. Tripleta Cerrada
Una tripleta cerrada es un conjunto de tres nodos donde todos los nodos están conectados entre sí, formando un ciclo o triángulo.

```python
# Crear un grafo dirigido
G_closed = nx.DiGraph()

# Añadir nodos y aristas para representar la tripleta cerrada
G_closed.add_nodes_from(["A", "B", "C"])
G_closed.add_edges_from([("A", "B"), ("B", "C"), ("C", "A")])

# Dibujar el grafo
pos = nx.spring_layout(G_closed)
nx.draw(G_closed, pos, with_labels=True, node_color="lightgreen", font_weight="bold", node_size=700)
nx.draw_networkx_edge_labels(G_closed, pos, edge_labels={("A", "B"): "A→B", ("B", "C"): "B→C", ("C", "A"): "C→A"})
plt.title("Tripleta Cerrada (A→B, B→C, y C→A)")
plt.show()
```

3. Tripleta Abierta con un Nodo No Conectado a los Otros Dos
Este es otro tipo de tripleta abierta donde uno de los nodos no está conectado a los otros dos nodos conectados entre sí.

```python
# Crear un grafo dirigido
G_open_variant = nx.DiGraph()

# Añadir nodos y aristas para representar la tripleta abierta variante
G_open_variant.add_nodes_from(["A", "B", "C"])
G_open_variant.add_edges_from([("A", "B")])  # Solo A está conectado a B, y C está desconectado

# Dibujar el grafo
pos = nx.spring_layout(G_open_variant)
nx.draw(G_open_variant, pos, with_labels=True, node_color="lightcoral", font_weight="bold", node_size=700)
nx.draw_networkx_edge_labels(G_open_variant, pos, edge_labels={("A", "B"): "A→B"})
plt.title("Tripleta Abierta Variante (A→B, y C no conectado)")
plt.show()
```

4. Tripleta con Un Nodo Central Conectado a Dos Nodos (Estructura de 'Y')
Este tipo de tripleta tiene un nodo central que está conectado a otros dos nodos, pero los dos nodos extremos no están conectados entre sí.

```python
# Crear un grafo dirigido
G_Y_shape = nx.DiGraph()

# Añadir nodos y aristas para representar la estructura de 'Y'
G_Y_shape.add_nodes_from(["A", "B", "C"])
G_Y_shape.add_edges_from([("B", "A"), ("B", "C")])  # B es el nodo central

# Dibujar el grafo
pos = nx.spring_layout(G_Y_shape)
nx.draw(G_Y_shape, pos, with_labels=True, node_color="violet", font_weight="bold", node_size=700)
nx.draw_networkx_edge_labels(G_Y_shape, pos, edge_labels={("B", "A"): "B→A", ("B", "C"): "B→C"})
plt.title("Tripleta en Forma de 'Y' (B→A y B→C, pero A y C no conectados)")
plt.show()
```

```python
from IPython.display import display, HTML

# Contenido HTML para mostrar en Google Colab
html_content = """
<!DOCTYPE html>
<html lang="es">
<head>
    <meta charset="UTF-8">
    <title>Búsqueda de Motivos en GraphFrame</title>
    <style>
        body {
            font-family: 'Arial', sans-serif;
            line-height: 1.6;
            padding: 20px;
            color: #333;
            background-color: #f5f5f5;
        }
        h1 {
            color: #0066cc;
            text-align: center;
        }
        h2 {
            color: #0056b3;
            margin-top: 20px;
        }
        ul {
            margin-left: 20px;
        }
        p {
            margin-bottom: 10px;
        }
        .collapsible {
            background-color: #0066cc;
            color: white;
            cursor: pointer;
            padding: 10px;
            width: 100%;
            text-align: left;
            outline: none;
            font-size: 1em;
            border: none;
            border-radius: 5px;
            margin-top: 10px;
            transition: background-color 0.3s;
        }
        .collapsible:hover {
            background-color: #0056b3;
        }
        .content {
            padding: 0 18px;
            display: none;
            overflow: hidden;
            background-color: #e0e7ef;
            border-radius: 5px;
            margin-bottom: 10px;
        }
    </style>
</head>
<body>
    <h1>Guía de Búsqueda de Motivos en GraphFrame</h1>
    <p>La búsqueda de motivos es una técnica utilizada para identificar patrones o estructuras específicas dentro de un grafo. Estos patrones, denominados motivos, representan subestructuras con formas específicas como bucles, cadenas o estrellas. La búsqueda de motivos es útil para analizar relaciones, dependencias y estructuras recurrentes, especialmente en redes sociales, redes biológicas y otros sistemas complejos.</p>

    <button class="collapsible">1. Conceptos Básicos del DSL de GraphFrame para la Búsqueda de Motivos</button>
    <div class="content">
        <p><strong>Vértices y Aristas:</strong> En un grafo, los vértices (nodos) representan entidades, y las aristas representan las relaciones entre ellos. En el DSL de GraphFrame:</p>
        <ul>
            <li>Los vértices se denotan con paréntesis, por ejemplo, <code>(a)</code>, donde "a" es el nombre asignado al vértice.</li>
            <li>Las aristas se denotan con corchetes, por ejemplo, <code>[e]</code>, donde "e" es el nombre de la arista.</li>
        </ul>
        <p><strong>Patrones como Secuencias de Aristas:</strong> Un motivo se describe mediante secuencias de aristas que conectan vértices. Por ejemplo:</p>
        <pre>"(a)-[e]->(b)"</pre>
        <p>Esto describe una arista <code>e</code> desde el vértice <code>a</code> al vértice <code>b</code>.</p>
        <p><strong>Combinación de Patrones con Punto y Coma:</strong> Múltiples aristas pueden combinarse usando punto y coma (<code>;</code>) para crear estructuras más complejas.</p>
        <p>Ejemplo:</p>
        <pre>"(a)-[e]->(b); (b)-[e2]->(c)"</pre>
        <p>Este motivo describe dos aristas, formando un camino desde el vértice <code>a</code> hasta <code>b</code> y luego hasta <code>c</code>.</p>
    </div>

    <button class="collapsible">2. Uso de Elementos Nombrados para Consultas</button>
    <div class="content">
        <p><strong>Nombres de Elementos:</strong> Al asignar nombres a vértices y aristas, puedes rastrear elementos específicos en varias partes de un patrón.</p>
        <p>Ejemplo:</p>
        <pre>"(a)-[e]->(b); (b)-[e2]->(a)"</pre>
        <p>Este motivo busca ciclos entre dos vértices, <code>a</code> y <code>b</code>, donde hay aristas en ambas direcciones.</p>
        <p><strong>Estructura del Resultado:</strong> Cuando ejecutas una consulta de búsqueda de motivos, GraphFrames devuelve un DataFrame en el que cada columna representa un elemento nombrado del patrón. Esto permite obtener información detallada sobre los vértices y aristas involucrados en el motivo.</p>
    </div>

    <button class="collapsible">3. Características Clave del DSL para Flexibilidad en Consultas</button>
    <div class="content">
        <p><strong>Elementos Anónimos:</strong> A veces no necesitas nombrar todos los elementos en el patrón.</p>
        <p>Ejemplo:</p>
        <pre>"(a)-[]->(b)"</pre>
        <p>Aquí buscamos cualquier arista entre <code>a</code> y <code>b</code>, pero no necesitamos información sobre la arista en sí, así que la dejamos sin nombre.</p>
        <p><strong>Negación:</strong> Puedes especificar aristas que no deben existir.</p>
        <p>Ejemplo:</p>
        <pre>"(a)-[]->(b); !(b)-[]->(a)"</pre>
        <p>Esto busca una conexión unidireccional desde <code>a</code> hasta <code>b</code>, asegurando que no exista una arista de <code>b</code> a <code>a</code>.</p>
    </div>

    <button class="collapsible">4. Restricciones y Reglas para Consultas de Motivos en GraphFrame</button>
    <div class="content">
        <ul>
            <li><strong>Patrones Anónimos Prohibidos:</strong> Los patrones no pueden consistir enteramente en elementos anónimos, como <code>()-[]->()</code> o <code>!()-[]->()</code>, ya que estos no proporcionan información útil.</li>
            <li><strong>Sin Aristas Nombradas en Negaciones:</strong> No está permitido nombrar aristas en términos negados, ya que esas aristas no aparecerían en los resultados. Ejemplo: <code>!(a)-[ab]->(b)</code> es inválido, pero <code>!(a)-[]->(b)</code> es permitido.</li>
        </ul>
    </div>

    <button class="collapsible">5. Aplicación de Filtros para Consultas Complejas</button>
    <div class="content">
        <p><strong>Filtrado por Atributos:</strong> Una vez que se devuelve el DataFrame de resultados, se pueden aplicar filtros adicionales para refinar la búsqueda. Esto es particularmente útil cuando deseas asegurarte de que los vértices o aristas en un motivo sean distintos.</p>
        <p>Ejemplo:</p>
        <pre>resultDataFrame.filter("a.id != c.id")</pre>
        <p>Este filtro garantiza que <code>a</code> y <code>c</code> se refieran a vértices distintos en el resultado.</p>
        <p><strong>Filas Duplicadas:</strong> Algunos motivos pueden devolver filas duplicadas. Por ejemplo, al consultar <code>(u)-[]->()</code> se encuentra cada arista desde un vértice <code>u</code> y puede devolver varias filas si múltiples aristas conectan <code>u</code> con otros vértices.</p>
    </div>

    <button class="collapsible">6. Ejemplo de Uso de la Búsqueda de Motivos en GraphFrame</button>
    <div class="content">
        <p>Supongamos que queremos encontrar un triángulo en un grafo de red social, donde cada vértice representa una persona y cada arista representa una amistad. Podríamos usar la siguiente consulta:</p>
        <pre>graph.find("(a)-[e1]->(b); (b)-[e2]->(c); (c)-[e3]->(a)")</pre>
        <p>Esto buscará grupos de tres personas donde cada persona es amiga de la siguiente en un patrón circular, formando un triángulo. El DataFrame resultante tendrá las columnas <code>a</code>, <code>b</code>, <code>c</code>, <code>e1</code>, <code>e2</code> y <code>e3</code>, representando los vértices y aristas de cada triángulo encontrado en el grafo.</p>
    </div>

    <script>
        var coll = document.getElementsByClassName("collapsible");
        for (var i = 0; i < coll.length; i++) {
            coll[i].addEventListener("click", function() {
                this.classList.toggle("active");
                var content = this.nextElementSibling;
                if (content.style.display === "block") {
                    content.style.display = "none";
                } else {
                    content.style.display = "block";
                }
            });
        }
    </script>
</body>
</html>
"""

# Mostrar el contenido en Google Colab
display(HTML(html_content))
```

```python
# Instalar NetworkX y Matplotlib
!pip install networkx matplotlib

import networkx as nx
import matplotlib.pyplot as plt

# Crear un grafo dirigido
G = nx.DiGraph()

# Agregar vértices
vertices = [("1", "Alice"), ("2", "Bob"), ("3", "Charlie"), ("4", "David"), ("5", "Eve")]
G.add_nodes_from([(id, {"name": name}) for id, name in vertices])

# Agregar aristas
edges = [
    ("1", "2"),
    ("2", "3"),
    ("3", "1"),
    ("3", "4"),
    ("4", "5"),
    ("5", "3")
]
G.add_edges_from(edges)

# Convertir a grafo no dirigido para detectar triángulos
UG = G.to_undirected()

# Detectar triángulos utilizando la función `triangles`
triangles = {node: nx.triangles(UG, node) for node in UG.nodes()}
print("Número de triángulos por nodo:")
for node, count in triangles.items():
    print(f"Vértice {node}: {count} triángulos")

# Detectar todos los triángulos únicos
triangles_unique = [cycle for cycle in nx.cycle_basis(UG) if len(cycle) == 3]
print("Triángulos detectados:")
for tri in triangles_unique:
    names = [G.nodes[node]['name'] for node in tri]
    print(tri, names)

# Graficar el grafo y resaltar los triángulos
plt.figure(figsize=(10, 8))
pos = nx.spring_layout(G)  # Disposición de nodos

# Dibujar todos los nodos y aristas del grafo
nx.draw(G, pos, with_labels=True, labels=nx.get_node_attributes(G, 'name'), node_size=700, node_color='lightblue', edge_color='gray', font_size=10, font_weight='bold')

# Resaltar triángulos detectados
for tri in triangles_unique:
    # Crear subgrafo con el triángulo
    triangle_edges = [(tri[i], tri[(i + 1) % 3]) for i in range(3)]
    nx.draw_networkx_edges(G, pos, edgelist=triangle_edges, width=2, edge_color='red')  # Aristas en rojo para resaltar triángulos
    nx.draw_networkx_nodes(G, pos, nodelist=tri, node_color='orange')  # Nodos en naranja

# Mostrar el grafo con los triángulos resaltados
plt.title("Grafo con Triángulos Detectados")
plt.show()
```

```python
# Paso 1: Instalar PySpark
!pip install pyspark

# Paso 2: Importar las librerías necesarias y configurar Spark
from pyspark.sql import SparkSession
from pyspark.sql.functions import col

# Inicializar la sesión de Spark
spark = SparkSession.builder \
    .appName("PySparkNetworkAnalysis") \
    .getOrCreate()

# Paso 3: Definir los datos de vértices y aristas
# Vértices
vertices = spark.createDataFrame([
    ("1", "Alice"),
    ("2", "Bob"),
    ("3", "Charlie"),
    ("4", "David"),
    ("5", "Eve")
], ["id", "name"])

# Aristas
edges = spark.createDataFrame([
    ("1", "2", "friend"),
    ("2", "3", "friend"),
    ("3", "1", "friend"),
    ("3", "4", "friend"),
    ("4", "5", "friend"),
    ("5", "3", "friend")
], ["src", "dst", "relationship"])

# Mostrar los vértices y aristas
print("Vértices:")
vertices.show()

print("Aristas:")
edges.show()

# Paso 4: Detectar triángulos utilizando joins
# Un triángulo se forma si (A -> B), (B -> C), y (C -> A) existen.

# Alias para las aristas
e1 = edges.alias("e1")
e2 = edges.alias("e2")
e3 = edges.alias("e3")

# Primer join: e1.dst = e2.src
joined1 = e1.join(e2, col("e1.dst") == col("e2.src")) \
           .select(
               col("e1.src").alias("A"),
               col("e1.dst").alias("B"),
               col("e2.dst").alias("C")
           )

# Segundo join: C -> A
triangles = joined1.join(e3, (col("C") == col("e3.src")) & (col("A") == col("e3.dst"))) \
                   .select("A", "B", "C") \
                   .distinct()

# Paso 5: Mostrar los triángulos encontrados
print("Triángulos detectados (incluyendo ciclos inválidos):")
triangles.show()

# Paso 6: Filtrar triángulos con tres nodos distintos
triangles_unique = triangles.filter(
    (col("A") != col("B")) &
    (col("B") != col("C")) &
    (col("C") != col("A"))
).distinct()

print("Triángulos detectados con nodos únicos:")
triangles_unique.show()

# Paso 7: Unir con los nombres de los vértices para mayor claridad
triangles_named = triangles_unique \
    .join(vertices, triangles_unique.A == vertices.id) \
    .withColumnRenamed("name", "A_name") \
    .join(vertices, triangles_unique.B == vertices.id) \
    .withColumnRenamed("name", "B_name") \
    .join(vertices, triangles_unique.C == vertices.id) \
    .withColumnRenamed("name", "C_name") \
    .select("A", "A_name", "B", "B_name", "C", "C_name")

print("Triángulos con nombres de vértices (nodos únicos):")
triangles_named.show()

# Paso 8: Detener la sesión de Spark
spark.stop()
```
