```python
from IPython.display import display, HTML

# Contenido HTML con comparación de BFS y DFS
html_content = """
<!DOCTYPE html>
<html lang="es">
<head>
    <meta charset="UTF-8">
    <meta name="viewport" content="width=device-width, initial-scale=1.0">
    <title>Comparación entre BFS y DFS</title>
    <link href="https://fonts.googleapis.com/css2?family=Roboto:wght@400;700&display=swap" rel="stylesheet">
    <style>
        body {
            font-family: 'Roboto', Arial, sans-serif;
            line-height: 1.8;
            margin: 20px;
            padding: 20px;
            background-color: #f9f9f9;
            color: #333;
        }
        h1, h2, h3 {
            color: #2c3e50;
        }
        h1 {
            font-size: 2.5em;
            text-align: center;
            margin-bottom: 20px;
        }
        h2 {
            font-size: 2em;
            margin-top: 40px;
            color: #2980b9;
        }
        h3 {
            font-size: 1.6em;
            margin-top: 30px;
            color: #2980b9;
        }
        p {
            font-size: 1.2em;
            margin-bottom: 15px;
        }
        ul {
            list-style-type: square;
            margin-left: 30px;
            margin-bottom: 20px;
        }
        pre {
            background-color: #f0f0f0;
            padding: 10px;
            border-radius: 5px;
            margin-bottom: 20px;
            font-family: monospace;
        }
        code {
            background-color: #f0f0f0;
            padding: 2px 5px;
            border-radius: 3px;
        }
        .highlight {
            background-color: #ffeb3b;
            padding: 2px 5px;
            border-radius: 3px;
            font-weight: bold;
        }
    </style>
</head>
<body>

    <h1>Comparación entre BFS y DFS</h1>

    <p><strong>BFS (Breadth-First Search)</strong> y <strong>DFS (Depth-First Search)</strong> son dos algoritmos fundamentales para recorrer o buscar nodos en un grafo o árbol. Ambos tienen enfoques diferentes para explorar los nodos, lo que afecta sus características y aplicaciones. A continuación, te ofrezco una comparación detallada.</p>

    <h2>1. Estrategia de Exploración</h2>
    <ul>
        <li><strong>BFS (Búsqueda en Anchura):</strong> Explora por niveles, comenzando desde el nodo raíz y explorando todos los nodos vecinos antes de pasar a los vecinos de los vecinos. Utiliza una cola (<code>FIFO</code>).</li>
        <li><strong>DFS (Búsqueda en Profundidad):</strong> Explora en profundidad, eligiendo un camino lo más profundo posible antes de retroceder. Utiliza una pila (<code>LIFO</code>) o se implementa de forma recursiva.</li>
    </ul>

    <h2>2. Estructura de Datos Utilizada</h2>
    <ul>
        <li><strong>BFS:</strong> Usa una cola (<code>FIFO</code>).</li>
        <li><strong>DFS:</strong> Usa una pila (<code>LIFO</code>) o la pila de recursión del sistema.</li>
    </ul>

    <h2>3. Recorrido y Orden de Visita</h2>
    <p><strong>BFS:</strong> Explora por capas. El primer nodo que encuentra el destino es el camino más corto en términos de número de aristas.</p>
    <p><strong>DFS:</strong> Explora en profundidad y puede no encontrar el camino más corto.</p>

    <pre>BFS: A → B → C → D → E → F</pre>
    <pre>DFS: A → B → G → H → C → D → E → F</pre>

    <h2>4. Complejidad Temporal y Espacial</h2>
    <ul>
        <li><strong>BFS:</strong>
            <ul>
                <li>Tiempo: <code>O(V + E)</code></li>
                <li>Espacio: <code>O(V)</code></li>
            </ul>
        </li>
        <li><strong>DFS:</strong>
            <ul>
                <li>Tiempo: <code>O(V + E)</code></li>
                <li>Espacio: <code>O(V)</code> iterativo, <code>O(d)</code> recursivo</li>
            </ul>
        </li>
    </ul>

    <h2>5. Encontrar el Camino Más Corto</h2>
    <ul>
        <li><strong>BFS:</strong> Encuentra el camino más corto en términos de número de aristas.</li>
        <li><strong>DFS:</strong> No garantiza encontrar el camino más corto en grafos no ponderados.</li>
    </ul>

    <h2>6. Detección de Ciclos</h2>
    <ul>
        <li><strong>BFS:</strong> Detecta ciclos si se encuentra un vecino ya visitado (excepto el padre).</li>
        <li><strong>DFS:</strong> Detecta ciclos si se revisita un nodo que ya está en la pila de recursión.</li>
    </ul>

    <h2>7. Aplicaciones Típicas</h2>
    <h3>BFS:</h3>
    <ul>
        <li>Búsqueda de caminos más cortos en grafos no ponderados.</li>
        <li>Problemas de conectividad.</li>
        <li>Análisis de niveles en grafos.</li>
        <li>Modelos de difusión (virus, información).</li>
    </ul>

    <h3>DFS:</h3>
    <ul>
        <li>Detección de ciclos.</li>
        <li>Ordenamiento topológico.</li>
        <li>Encontrar componentes conectados.</li>
        <li>Problemas de retroceso (backtracking).</li>
    </ul>

    <h2>8. Ejemplo Visual</h2>
    <pre>BFS: A → B → C → G → H → D → E → F</pre>
    <pre>DFS: A → B → G → H → C → D → E → F</pre>

    <h2>9. Ventajas y Desventajas</h2>
    <h3>BFS:</h3>
    <ul>
        <li><strong>Ventajas:</strong> Encuentra el camino más corto. Ideal para búsquedas cercanas al origen.</li>
        <li><strong>Desventajas:</strong> Puede consumir más memoria. Ineficiente en grafos profundos.</li>
    </ul>

    <h3>DFS:</h3>
    <ul>
        <li><strong>Ventajas:</strong> Menor consumo de memoria. Ideal para recorridos completos y problemas de backtracking.</li>
        <li><strong>Desventajas:</strong> No garantiza el camino más corto. Puede requerir mucha memoria en la versión recursiva para grafos profundos.</li>
    </ul>

</body>
</html>
"""

# Mostrar el contenido HTML en Colab
display(HTML(html_content))
```

```python
from IPython.display import display, HTML

# Contenido HTML con explicación de DFS Tree
html_content = """
<!DOCTYPE html>
<html lang="es">
<head>
    <meta charset="UTF-8">
    <meta name="viewport" content="width=device-width, initial-scale=1.0">
    <title>DFS (Depth-First Search Tree)</title>
    <link href="https://fonts.googleapis.com/css2?family=Roboto:wght@400;700&display=swap" rel="stylesheet">
    <style>
        body {
            font-family: 'Roboto', Arial, sans-serif;
            line-height: 1.8;
            margin: 20px;
            padding: 20px;
            background-color: #f9f9f9;
            color: #333;
        }
        h1, h2, h3 {
            color: #2c3e50;
        }
        h1 {
            font-size: 2.5em;
            text-align: center;
            margin-bottom: 20px;
        }
        h2 {
            font-size: 2em;
            margin-top: 40px;
            color: #2980b9;
        }
        p {
            font-size: 1.2em;
            margin-bottom: 15px;
        }
        code {
            background-color: #f0f0f0;
            padding: 2px 5px;
            border-radius: 3px;
        }
        ul {
            list-style-type: square;
            margin-left: 30px;
            margin-bottom: 20px;
        }
        .highlight {
            background-color: #ffeb3b;
            padding: 2px 5px;
            border-radius: 3px;
            font-weight: bold;
        }
    </style>
</head>
<body>

    <h1>DFS (Depth-First Search Tree)</h1>

    <p>El <strong>árbol de búsqueda en profundidad (DFS)</strong> es un tipo de recorrido de grafos que explora lo más profundo posible en cada rama antes de retroceder. Esto significa que sigue un camino hasta llegar a un vértice sin vecinos no visitados, y luego retrocede (o "retrocede") para explorar otras rutas. El resultado es un árbol que refleja el orden en que los nodos fueron descubiertos durante la exploración.</p>

    <h2>Pasos para construir un DFS Tree:</h2>
    <ul>
        <li><strong>Comenzar en un vértice inicial:</strong> Se selecciona un nodo de partida, que será la raíz del árbol.</li>
        <li><strong>Explorar hacia lo más profundo:</strong> El algoritmo elige un vecino no visitado y lo visita, repitiendo el proceso con el nuevo nodo.</li>
        <li><strong>Retroceder si no hay más vecinos:</strong> Si se llega a un nodo sin vecinos no visitados, se retrocede al nodo anterior para seguir explorando otras rutas no exploradas.</li>
        <li><strong>Repetir hasta que todos los nodos hayan sido visitados:</strong> Se continúa hasta que todos los nodos conectados hayan sido explorados.</li>
    </ul>

    <h2>Características:</h2>
    <ul>
        <li>El DFS produce un árbol de búsqueda en profundidad que muestra cómo los nodos fueron descubiertos.</li>
        <li>Se puede aplicar en grafos dirigidos y no dirigidos.</li>
        <li>Tiende a explorar caminos largos antes de buscar rutas alternativas más cortas.</li>
    </ul>

    <h2>Ejemplo:</h2>
    <p>Imagina un grafo no dirigido:</p>
    <pre>
    A
   / \\
  B   C
 /     \\
D       E
         \\
          F
    </pre>

    <p>Si empezamos el DFS en el nodo <code>A</code>, el recorrido podría ser el siguiente:</p>
    <ul>
        <li>Desde <code>A</code>, se explora <code>B</code>.</li>
        <li>Desde <code>B</code>, se explora <code>D</code>.</li>
        <li>Al no tener más vecinos no visitados, retrocede a <code>B</code> y luego a <code>A</code>.</li>
        <li>Desde <code>A</code>, se explora <code>C</code>.</li>
        <li>Desde <code>C</code>, se explora <code>E</code>.</li>
        <li>Desde <code>E</code>, se explora <code>F</code>.</li>
    </ul>

    <p>El árbol DFS resultante será:</p>
    <pre>
    A
   / \\
  B   C
 /     \\
D       E
         \\
          F
    </pre>

    <h2>Aplicaciones:</h2>
    <ul>
        <li>Detección de ciclos en grafos.</li>
        <li>Encontrar componentes fuertemente conectados en grafos dirigidos.</li>
        <li>Problemas de conectividad en grafos no dirigidos.</li>
        <li>Búsqueda en laberintos o problemas de enrutamiento.</li>
    </ul>

    <p>El <strong>DFS</strong> es muy útil cuando necesitas explorar caminos específicos y obtener información sobre la estructura interna del grafo o la conectividad de sus nodos.</p>

</body>
</html>
"""

# Mostrar el contenido HTML en Colab
display(HTML(html_content))
```



### Preguntas y Respuestas

---

**1. ¿Qué es BFS (Breadth-First Search) y cómo funciona?**

> **Respuesta**: BFS (Búsqueda en Anchura) es un algoritmo de búsqueda en grafos que explora todos los nodos vecinos de un nodo inicial antes de avanzar a los vecinos de esos vecinos. Usa una cola (FIFO) para procesar cada nodo en el orden en que fue descubierto, asegurando un recorrido por niveles. Este enfoque es útil para encontrar el camino más corto en términos de número de aristas en grafos no ponderados.

---

**2. ¿Cómo se diferencia DFS (Depth-First Search) de BFS?**

> **Respuesta**: DFS (Búsqueda en Profundidad) explora lo más profundo posible en una rama antes de retroceder y explorar otras rutas. A diferencia de BFS, que usa una cola, DFS emplea una pila (LIFO) o se implementa de forma recursiva. DFS es útil en aplicaciones que requieren explorar caminos completos, como en problemas de retroceso (backtracking).

---

**3. ¿Qué estructura de datos utiliza BFS y por qué?**

> **Respuesta**: BFS utiliza una cola (FIFO) para almacenar los nodos a explorar. La cola permite que los nodos se procesen en el orden en que fueron descubiertos, lo que garantiza un recorrido por niveles desde el nodo inicial hasta todos sus vecinos de manera ordenada.

---

**4. ¿Qué estructura de datos utiliza DFS y por qué?**

> **Respuesta**: DFS usa una pila (LIFO) o la pila de recursión del sistema, lo que permite que el algoritmo explore en profundidad. Esta estructura es adecuada para DFS, ya que permite al algoritmo seguir un camino específico hasta el final antes de retroceder y explorar caminos alternativos.

---

**5. ¿Cuál es la complejidad temporal y espacial de BFS?**

> **Respuesta**: La complejidad temporal de BFS es \(O(V + E)\), donde \(V\) es el número de vértices y \(E\) el número de aristas, ya que explora todos los nodos y aristas en el peor caso. En cuanto a complejidad espacial, BFS requiere \(O(V)\) debido al almacenamiento de todos los nodos en la cola.

---

**6. ¿Cuándo es preferible utilizar DFS sobre BFS?**

> **Respuesta**: DFS es preferible en problemas de retroceso (backtracking), como en la resolución de laberintos o puzzles, donde se necesita explorar caminos completos. También es útil para la detección de ciclos en grafos y el análisis de componentes conectados. DFS tiende a ser más eficiente en términos de memoria para grafos menos profundos.

---

**7. ¿Cómo garantiza BFS encontrar el camino más corto en un grafo no ponderado?**

> **Respuesta**: En un grafo no ponderado, BFS explora todos los nodos a una distancia dada del nodo inicial antes de avanzar a una mayor distancia. Esto asegura que el primer camino encontrado hacia un nodo es el más corto en términos de número de aristas, ya que BFS procesa cada capa de nodos de manera ordenada.

---

**8. ¿Cómo puede DFS detectar ciclos en un grafo?**

> **Respuesta**: DFS detecta ciclos si, durante el proceso de exploración, vuelve a visitar un nodo que ya está en la pila de recursión. Esto indica la presencia de un ciclo, ya que existe un camino de regreso al nodo sin haber completado el recorrido.

---

**9. ¿Cuáles son las aplicaciones típicas de BFS?**

> **Respuesta**: BFS se utiliza para la búsqueda de caminos más cortos en grafos no ponderados, problemas de conectividad, análisis de niveles de grafos y modelos de difusión (como la propagación de virus o información). Es útil en cualquier contexto donde sea importante explorar un grafo de manera organizada y por niveles.

---

**10. ¿Cuáles son las principales ventajas y desventajas de DFS?**

> **Respuesta**: La principal ventaja de DFS es su bajo consumo de memoria en la versión iterativa y su eficacia en recorridos completos y problemas de retroceso. Sin embargo, DFS no garantiza encontrar el camino más corto en grafos no ponderados y puede requerir mucha memoria en la implementación recursiva para grafos muy profundos.

---
