```python
# Importar las bibliotecas necesarias
import matplotlib.pyplot as plt
import networkx as nx
import numpy as np
from matplotlib.animation import FuncAnimation, PillowWriter
from IPython.display import HTML
from google.colab import files

# Crear el grafo de Petersen
G = nx.petersen_graph()

# Definir posiciones de los nodos
pos = {
    0: (0, 1.5),    # Nodo 1
    1: (-1.4, 0.5), # Nodo 2
    2: (-0.8, -1.2),# Nodo 3
    3: (0.8, -1.2), # Nodo 4
    4: (1.4, 0.5),  # Nodo 5
    5: (0, 0.6),    # Nodo 6
    6: (-0.5, 0.2), # Nodo 7
    7: (-0.4, -0.7),# Nodo 8
    8: (0.4, -0.7), # Nodo 9
    9: (0.5, 0.2)   # Nodo 10
}

# Definir el nodo de inicio para el recorrido BFS y los nodos en el orden de BFS
start_node = 0
bfs_edges = list(nx.bfs_edges(G, start_node))
bfs_nodes = [start_node] + [v for u, v in bfs_edges]  # Orden de nodos en BFS

# Configurar el gráfico
fig, ax = plt.subplots(figsize=(8, 8))
plt.axis('off')  # Ocultar los ejes

# Inicializar la lista de colores de los nodos
node_colors = ['gray'] * len(G.nodes)

# Función de actualización para la animación
def update(frame):
    # Reiniciar colores de los nodos a gris (no visitados)
    node_colors = ['gray'] * len(G.nodes)

    # Colorear nodos visitados hasta el frame actual en azul
    for i in range(frame):
        node_colors[bfs_nodes[i]] = 'blue'  # Nodos ya visitados

    # Colorear el nodo actual en rojo
    if frame < len(bfs_nodes):
        node_colors[bfs_nodes[frame]] = 'red'  # Nodo en proceso de visita

    # Limpiar el gráfico y volver a dibujar
    ax.clear()
    nx.draw_networkx_nodes(G, pos, node_color=node_colors, node_size=700, ax=ax)
    nx.draw_networkx_labels(G, pos, font_color='white', font_size=12, ax=ax)
    nx.draw_networkx_edges(G, pos, edge_color='black', width=2, ax=ax)
    plt.title(f"Recorrido BFS - Paso {frame}")

# Crear la animación
ani = FuncAnimation(fig, update, frames=len(bfs_nodes), interval=1000, repeat=False)

# Guardar la animación como GIF
gif_path = "/content/bfs_animacion.gif"
ani.save(gif_path, writer=PillowWriter(fps=1))
print("GIF guardado!")

# Guardar la animación como MP4
mp4_path = "/content/bfs_animacion.mp4"
ani.save(mp4_path, writer='ffmpeg', fps=1)
print("MP4 guardado!")

# Mostrar el GIF en el notebook
HTML(f'<img src="{gif_path}" width="600">')

# Descargar archivos
files.download(gif_path)
files.download(mp4_path)
```

```python
# Importar las bibliotecas necesarias
import matplotlib.pyplot as plt
import networkx as nx
from matplotlib.animation import FuncAnimation, PillowWriter
from IPython.display import HTML
from google.colab import files

# Crear el grafo de Petersen
G = nx.petersen_graph()

# Definir posiciones de los nodos manualmente para una disposición más clara
pos = {
    0: (0, 1.5),    # Nodo 1
    1: (-1.4, 0.5), # Nodo 2
    2: (-0.8, -1.2),# Nodo 3
    3: (0.8, -1.2), # Nodo 4
    4: (1.4, 0.5),  # Nodo 5
    5: (0, 0.6),    # Nodo 6
    6: (-0.5, 0.2), # Nodo 7
    7: (-0.4, -0.7),# Nodo 8
    8: (0.4, -0.7), # Nodo 9
    9: (0.5, 0.2)   # Nodo 10
}

# Definir el nodo de inicio para el recorrido BFS y los nodos en el orden de BFS
start_node = 0
bfs_edges = list(nx.bfs_edges(G, start_node))
bfs_nodes = [start_node] + [v for u, v in bfs_edges]  # Orden de nodos en BFS

# Configurar el gráfico
fig, ax = plt.subplots(figsize=(8, 8))
plt.axis('off')  # Ocultar los ejes

# Inicializar la lista de colores de los nodos
node_colors = ['gray'] * len(G.nodes)

# Función de actualización para la animación
def update(frame):
    # Reiniciar colores de los nodos a gris (no visitados)
    node_colors = ['gray'] * len(G.nodes)

    # Colorear nodos visitados hasta el frame actual en azul
    for i in range(frame % len(bfs_nodes)):  # Uso de módulo para ciclo continuo
        node_colors[bfs_nodes[i]] = 'blue'  # Nodos ya visitados

    # Colorear el nodo actual en rojo
    current_node = bfs_nodes[frame % len(bfs_nodes)]
    node_colors[current_node] = 'red'  # Nodo en proceso de visita

    # Limpiar el gráfico y volver a dibujar
    ax.clear()
    nx.draw_networkx_nodes(G, pos, node_color=node_colors, node_size=700, ax=ax)
    nx.draw_networkx_labels(G, pos, font_color='white', font_size=12, ax=ax)
    nx.draw_networkx_edges(G, pos, edge_color='black', width=2, ax=ax)
    plt.title(f"Recorrido BFS - Paso {frame % len(bfs_nodes)}")

# Crear la animación con repetición continua y un ciclo sin fin en los nodos
ani = FuncAnimation(fig, update, frames=len(bfs_nodes) * 2, interval=500, repeat=True)

# Guardar la animación como MP4 con movimiento continuo
mp4_path = "/content/bfs_animacion_continua.mp4"
ani.save(mp4_path, writer='ffmpeg', fps=2)  # Ajuste de fps para movimiento continuo
print("MP4 guardado con movimiento continuo!")

# Mostrar el MP4 en el notebook
HTML(f'<video width="600" controls><source src="{mp4_path}" type="video/mp4"></video>')

# Descargar el archivo MP4
files.download(mp4_path)
```
